```python
import netCDF4 as nc
import matplotlib.pyplot as plt
import numpy as np
from salishsea_tools import viz_tools,tidetools

%matplotlib inline
```

```python
grid = '/data/nsoontie/MEOPAR/NEMO-forcing/grid/bathy_meter_SalishSea2.nc'
fB = nc.Dataset(grid)
bathy=fB.variables['Bathymetry'][:]
X = fB.variables['nav_lon'][:]
Y=fB.variables['nav_lat'][:]
```

```python
#Seymour Narrows
lat36 = 50.135820
lon36 = -125.353403

j36, i36 = tidetools.find_closest_model_point(lon36, lat36, X, Y, bathy)
print i36, j36

#Arran Rapids
lat37 = 50.416667
lon37 = -125.133333
j37, i37 = tidetools.find_closest_model_point(lon37,lat37,X, Y, bathy)
print i37, j37

#Thalweg 1300
thalweg = np.loadtxt('/data/dlatorne/MEOPAR/tools/bathymetry/thalweg_working.txt', dtype=int, unpack=True)
i13 = thalweg[1,1303]
j13 = thalweg[0,1303]
print i13, j13, X[j13,i13], Y[j13, i13]
```

```python
plt.pcolormesh(bathy)
plt.plot(i36,j36,'ro')
plt.plot(i13,j13,'ro')
plt.axis([110,140,750,790])
```

```python
ufil='/data/dlatorne/MEOPAR/SalishSea/nowcast/24jan15/SalishSea_1d_20150124_20150124_grid_U.nc'
fU = nc.Dataset(ufil)
U = fU.variables['vozocrtx']
U=np.ma.masked_values(U[0,0,:,:],0)

plt.pcolormesh(U)
plt.plot(i36,j36,'ro')
plt.plot(i13,j13,'go')
plt.axis([110,125,765,780])

print U[j36,i36]
print U[j36,i36-1]
print U[j13,i13]
print U[j13,i13-1]
```

```python
vfil='/data/dlatorne/MEOPAR/SalishSea/nowcast/24jan15/SalishSea_1d_20150124_20150124_grid_V.nc'
fV = nc.Dataset(vfil)
V = fV.variables['vomecrty']
V=np.ma.masked_values(V[0,0,:,:],0)

plt.pcolormesh(V)
plt.plot(i36,j36,'ro')
plt.plot(i13,j13,'go')
plt.axis([110,125,765,780])

print V[j36,i36]
print V[j36-1,i36]
print V[j13,i13]
print V[j13-1,i13]
```
