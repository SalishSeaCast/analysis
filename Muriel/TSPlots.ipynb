This notebook will load the temperature and salinity values of all the Bamfield paper stations and make TS plots

```python
import numpy as np
import matplotlib.pylab as plt
import matplotlib.colors as mcolors
import seaborn as sns

%matplotlib inline
```

```python
G1=np.loadtxt('G1TS', skiprows=1, unpack=True)
G2=np.loadtxt('G2TS', skiprows=1, unpack=True)
G3=np.loadtxt('G3TS', skiprows=1, unpack=True)
G4=np.loadtxt('G4TS', skiprows=1, unpack=True)
G5=np.loadtxt('G5TS', skiprows=1, unpack=True)
PO=np.loadtxt('POTS', skiprows=1, unpack=True)
WI=np.loadtxt('WITS', skiprows=1, unpack=True)
DI=np.loadtxt('DITS', skiprows=1, unpack=True)
```

```python
font = {'family' : 'Arial',
    'weight' : 'normal',
    'size'   : 18}
axisfont = {'family' : 'Arial',
    'weight' : 'light',
    'size'   : 14}
```

```python
fig, ax = plt.subplots(1,1,figsize=(18,8))
palette = sns.light_palette("green" , 6)
palette[1]
cmap1 = mcolors.ListedColormap(palette)
for G, i in zip([G1, G2, G3, G4, G5], np.arange(1,6)):
    plt.plot(G[1], G[0], '-', c=palette[i], label='G{}'.format(i))
ax.set_ylabel('Temprature (deg C)', **axisfont)
ax.set_xlabel('Salinity (PSU)', **axisfont)
ax.legend(loc=2)
ax.set_title('Water Masses Through the Grappler Inlet', **font)
```

```python
fig, ax = plt.subplots(1,1,figsize=(15,10))
cmap1 = mcolors.ListedColormap(palette)
name = ['Wizard Island', 'Diana Island', 'Pacific Ocean']
for loc, name in zip([WI, DI, PO], name):
    plt.plot(loc[1], loc[0], '.-', label='{}'.format(name))
ax.set_ylabel('Temprature (deg C)', **axisfont)
ax.set_xlabel('Salinity (PSU)', **axisfont)
ax.legend(loc=2)
ax.set_title('Water Masses at Other Stations', **font)
```
