This notebook will load the temperature and salinity values of all the Bamfield paper stations and make TS plots

```python
import numpy as np
import matplotlib.pylab as plt
import matplotlib.colors as mcolors
import seaborn as sns

%matplotlib inline
```

```python
G1=np.loadtxt('/home/mdunn/Documents/G1TS', skiprows=1, unpack=True)
G2=np.loadtxt('/home/mdunn/Documents/G2TS', skiprows=1, unpack=True)
G3=np.loadtxt('/home/mdunn/Documents/G3TS', skiprows=1, unpack=True)
G4=np.loadtxt('/home/mdunn/Documents/G4TS', skiprows=1, unpack=True)
G5=np.loadtxt('/home/mdunn/Documents/G5TS', skiprows=1, unpack=True)
PO=np.loadtxt('/home/mdunn/Documents/POTS', skiprows=1, unpack=True)
WI=np.loadtxt('/home/mdunn/Documents/WITS', skiprows=1, unpack=True)
DI=np.loadtxt('/home/mdunn/Documents/DITS', skiprows=1, unpack=True)
TC=np.loadtxt('/home/mdunn/Documents/TCTS', skiprows=1, unpack=True)
```

```python
font = {'family' : 'Arial',
    'weight' : 'normal',
    'size'   : 18}
axisfont = {'family' : 'Arial',
    'weight' : 'light',
    'size'   : 14}
```

```python
fig, ax = plt.subplots(1,1,figsize=(18,8))
palette = sns.light_palette("green" , 6)
palette[1]
cmap1 = mcolors.ListedColormap(palette)
for G, i in zip([G1, G2, G3, G4, G5], np.arange(1,6)):
    plt.plot(G[1], G[2], '-', c=palette[i], label='G{}'.format(i))
ax.tick_params(axis='both', which='major', labelsize=14)
ax.set_ylabel('Temprature (deg C)', **axisfont)
ax.set_xlabel('Salinity (PSU)', **axisfont)
ax.legend(loc=2)
ax.set_title('Water Masses Through the Grappler Inlet', **font)
```
