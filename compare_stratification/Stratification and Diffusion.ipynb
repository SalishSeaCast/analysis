Comparison of first year of NEMO spin up with StratoGEM. Focus is on salinity. 

```python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
import netCDF4 as NC

from salishsea_tools import viz_tools
from salishsea_tools import bathy_tools
from salishsea_tools import nc_tools
from salishsea_tools import tidetools

from __future__ import division

import glob
import os
import datetime
import arrow
from scipy import interpolate as interp
```

# STRATOGEM Prep

```python

# Read from a file, the location of our stations
f = open("station_locations",'r')
data = f.readlines()
count = 0
stn = [None]*9; depth = [None]*9; strat_lats = np.zeros(9); strat_lons = np.zeros(9)
for line in data:
    word = line.split()
    if word[1] == 'Station:':
        stn[count] = word[2]
    if word[1] == 'Depth:':
        depth[count] = word[2]
    if word[1] == 'Latitude':
        strat_lats[count] = float(word[3])+float(word[4])/60.
    if word[1] == 'Longitude':
        strat_lons[count] = float(word[3])+float(word[4])/60.
        count = count + 1

```

```python
#       cruise code   date          yr  nu   directory     file           timecounter
cruises = {'0208': ('Oct 8, 2002', '02','08','8oct9oct',  '1d','20021008_20021011',1),
           '0209': ('Nov 5, 2002', '02','09','5nov14nov', '1d','20021105_20021114',0),
           '0210': ('Dec 4, 2002', '02','10','30nov9dec', '1d','20021130_20021209',4),
           '0301': ('Jan 8, 2003', '03','01','6jan10jan', '1d','20030106_20030110',2),
           '0302': ('Feb 5, 2003', '03','02','31jan9feb', '1d','20030131_20030209',5),
           '0303': ('Mar 5, 2003', '03','03','2mar11mar', '1d','20030302_20030311',3),
           '0305': ('Apr 2, 2003', '03','05','1apr10apr', '1d','20030401_20030410',1),
           '0308': ('Apr 23, 2003','03','08','21apr30apr','1d','20030421_20030430',2),
           '0309': ('May 28, 2003','03','09','21may30may','1d','20030521_20030530',7),
           '0314': ('Jul 3, 2003', '03','14','30jun9jul', '1d','20030630_20030709',3),
           '0315': ('Aug 14, 2003','03','15','9aug18aug', '1d','20030809_20030818',5)
           }
```

```python
def read_STRATOGEM(cruiseno):
    """Reads the data from a STRATOGEM cruise. Returns temp, sal, depth"""
    ss_depth = np.zeros((9,500)); ss_sal = np.zeros((9,500)); ss_temp = np.zeros((9,500))
    count = np.zeros((9), dtype=np.int)
    
    path = "/ocean/shared/SoG/ctd/{yr}-{nu}/5_bin/".format(yr=cruises[cruiseno][1],
                                                           nu=cruises[cruiseno][2])
    for fileno in range(10):
        filename = "SG{cruiseno}00{fileno}.cnv".format(cruiseno=cruiseno, fileno=fileno)
        try: 
            filehandle = open(os.path.join(path,filename),'rt')
        except IOError:
            break
        with filehandle as data:
            headerover = False
            for line in data:
                if headerover:
                    word = line.split()
                    ss_depth[i_stn,count[i_stn]] = float(word[1])
                    ss_sal[i_stn,count[i_stn]] = float(word[salkey])
                    ss_temp[i_stn,count[i_stn]] = float(word[tempkey])
                    count[i_stn] = count[i_stn] + 1
                elif line[0:5] == "*END*":
                    headerover = True
                else:
                    word = line.split()
                    if len(word) > 1:
                        if word[1] == "Station:":
                            ss_stn = word[2]
                            i_stn = stn.index(ss_stn)
                        elif word[1] == "name":
                            if word[5] == "Salinity":
                                salkey = int(word[2])
                            elif word[5] == "Temperature":
                                tempkey = int(word[2])
    return ss_depth, ss_sal, ss_temp
```

```python

bathy, X, Y = tidetools.get_SS_bathy_data()

path = '/ocean/dlatorne/MEOPAR/SalishSea/results/spin-up/'
files = glob.glob(path+'*/SalishSea_1d_*grid_T.nc')
files.sort(key=os.path.basename)

#load all files as if it where one dataset
tracers = NC.MFDataset(files)
count= tracers.variables['time_counter']
torig=nc_tools.time_origin(tracers)
times = []
for ind in np.arange(count.shape[0]):
    times.append((torig + datetime.timedelta(seconds=int(count[ind]))))
times=np.array(times)
ds = tracers.variables['deptht']

sal = tracers.variables['vosaline']
```

```python
def interpolate_depth(sal,depth_array,depth_new):
    """ interpolates a salinity field given at ds to a desire depth """
    mu=sal==0
    salo=np.ma.array(sal,mask=mu)
    mu=depth_array==0
    depth_arrayo=np.ma.array(depth_array,mask=mu)
    f= interp.interp1d(depth_arrayo,salo)
    sali = f(depth_new)
    return sali
```

# Plotting

```python
def compare_NEMO_STRATOGEM(depth,stnum,ax,smin,smax):
    station_name = stn[stnum]
    dates=[]
    keys = ('0301','0302','0303','0305','0308','0309','0314','0315')
    
    S_sals = np.zeros(len(keys)); N_sals=[]
    count=0
    for key in ('0301','0302','0303','0305','0308','0309','0314','0315'):
        date = datetime.datetime.strptime(cruises[key][0], '%b %d, %Y')
        dates.append(date)
        
        #stratogem
        ss_depth, ss_sal, ss_temp = read_STRATOGEM(key)
        mu=ss_sal[stnum,:]==0
        ss = ss_sal[stnum,ss_depth[stnum,:]==depth]
        if not ss:
            ss='Nan'
        S_sals[count]=ss

        #NEMO
        index = int(datetime.datetime.strftime(date,'%j'))-1
        [j,i] = tidetools.find_closest_model_point(-strat_lons[stnum],strat_lats[stnum],X,Y,bathy)
        ss=interpolate_depth(sal[index,:,j,i],ds,depth)
        N_sals.append(ss)
        count=count+1
    dates=np.array(dates)
    N_sals=np.array(N_sals)
    
    ax.plot(dates,S_sals,'b-o',label='Stratogem')
    ax.plot(dates,N_sals,'g-*',label='NEMO')
    ax.set_title('Station ' + stn[stnum])
    ax.set_ylim([smin,smax])
    ax.set_ylabel('Salinity (PSU)?')
    ax.legend(loc=0)
```

```python
def plot_stations(depth,smin,smax):
    fig,axs = plt.subplots(3,3,figsize=(20,10))
    for st,ax in zip(np.arange(9),axs.flat):
        compare_NEMO_STRATOGEM(depth,st,ax,smin,smax)
```
