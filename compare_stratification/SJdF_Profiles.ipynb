```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import netCDF4 as nc
import os
import datetime
import glob

from salishsea_tools import viz_tools
from salishsea_tools import tidetools
from salishsea_tools import nc_tools
```

```python
def read_IOS_file(filename,i_stn):
    filehandle = open(filename,'rt')
    with filehandle as data:
            headerover = False
            for line in data:
                if headerover:
                    word = line.split()
                    ss_depth[i_stn,count[i_stn]] = float(word[0])
                    ss_sal[i_stn,count[i_stn]] = float(word[5])
                    ss_temp[i_stn,count[i_stn]] = float(word[1])
                    count[i_stn] = count[i_stn] + 1
                elif line[0:14] == "*END OF HEADER":
                    headerover = True
                else:
                    word = line.split()
                    if len(word) > 1:
                        if word[0] == "LATITUDE":
                            lat = float(word[2])+ float(word[3])/60
                            lats[i_stn]=lat
                        if word[0] == "LONGITUDE":
                            lon = -(float(word[2])+ float(word[3])/60)
                            lons[i_stn]=lon
                        if word[0] == "START":
                            string = word[4] +' ' + word[5]
                            time = datetime.datetime.strptime(string,'%Y/%m/%d %H:%M:%S.%f')
                            times.append(time)
    return lat,lon,time
```

```python
def compare_salinity(d1,d2,date,smin=28,smax=32):
    
    fig,ax=plt.subplots(1,1,figsize=(10,5))
    #Isolate model date
    tind = date.day - 21
    time_m=nc_tools.timestamp(gridT,tind)
    
    #Isolate modeldepth indices between depths observed d1 and observed d2
    dinds = np.where(np.logical_and(depth_m[:]>=d1,depth_m[:]<=d2))
    dinds = dinds[0]
    
    #Loop through obervations. For ones on date, plot salinity vs longitude for obs and model.
    sal_model=[];sal_obs=[]; lon_plotm=[]; lon_ploto=[]
    for n in range(times.shape[0]):
        if times[n].day == date.day:
            #observations data
            #extract depth range
            obs_dinds=np.where(np.logical_and(ss_depth[n,:]>=d1,ss_depth[n,:]<=d2))
            dz=np.array(ss_depth[n,obs_dinds]) #array of observed depths
            if dz.size ==0: #check there are any observations in this depth range
                obs_plot='NaN'
            else:
                #average
                obs_range=np.ptp(dz)
                print ' Observed depth range: {} {} m'.format(dz[0][0],dz[0][-1])
                obs_plot=np.trapz(ss_sal[n,obs_dinds],x=dz)/np.ptp(dz)
            sal_obs.append(obs_plot); lon_ploto.append(lons[n])
            # model data
            [j,i]=tidetools.find_closest_model_point(lons[n],lats[n],X,Y,bathy,allow_land=False)
            if not isinstance(j,list): #check in domain
                sal_plot=sal_m[tind,dinds,j,i]
                #mask to be safe
                mu=sal_plot==0; sal_plot=np.ma.array(sal_plot,mask=mu)
                dz=np.ma.array(depth_m[dinds],mask=mu); model_range=np.ptp(dz);
                #average
                print ' Model depth range: {} {} m'.format(depth_m[dinds[0]],depth_m[dinds[-1]])
                sal_plot=np.trapz(sal_plot,x=dz)/model_range
                sal_model.append(sal_plot);lon_plotm.append(lons[n])   
           

    #sort so plot in a line along longitude
    model_sort = [x for y, x in sorted(zip(lon_plotm, sal_model))]
    lon_plotm.sort()
    obs_sort = [x for y, x in sorted(zip(lon_ploto, sal_obs))]
    lon_ploto.sort()
    
    #plot
    ax.plot(lon_plotm,model_sort,'bo-',label='Model')
    ax.plot(lon_ploto,obs_sort,'rs-',label='Observation')
    ax.legend(loc=0)
    ax.set_ylabel('Salinity [psu]'); ax.set_xlabel('Longitude')
    ax.set_title('Salinity on {date} averaged over {d1} - {d2} m'.format(date=date.strftime('%Y/%m/%d'),d1=d1,d2=d2))
    #ax.set_ylim([smin,smax])
    ax.set_xlim([-125,-122.3])
    
    return fig
```
