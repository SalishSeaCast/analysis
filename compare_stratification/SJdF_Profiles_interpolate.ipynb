```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import netCDF4 as nc
import os
import datetime
import glob

from salishsea_tools import viz_tools
from salishsea_tools import tidetools
from salishsea_tools import nc_tools

from scipy import interpolate as interp
```

```python
def read_IOS_file(filename,i_stn):
    filehandle = open(filename,'rt')
    with filehandle as data:
            headerover = False
            for line in data:
                if headerover:
                    word = line.split()
                    ss_depth[i_stn,count[i_stn]] = float(word[0])
                    ss_sal[i_stn,count[i_stn]] = float(word[5])
                    ss_temp[i_stn,count[i_stn]] = float(word[1])
                    count[i_stn] = count[i_stn] + 1
                elif line[0:14] == "*END OF HEADER":
                    headerover = True
                else:
                    word = line.split()
                    if len(word) > 1:
                        if word[0] == "LATITUDE":
                            lat = float(word[2])+ float(word[3])/60
                            lats[i_stn]=lat
                        if word[0] == "LONGITUDE":
                            lon = -(float(word[2])+ float(word[3])/60)
                            lons[i_stn]=lon
                        if word[0] == "START":
                            string = word[4] +' ' + word[5]
                            time = datetime.datetime.strptime(string,'%Y/%m/%d %H:%M:%S.%f')
                            times.append(time)
    return lat,lon,time
```

Loop through the files to grab loations and times

```python
files =glob.glob('/ocean/sallen/allen/research/MEOPAR/IOS_2003_CTD/2003-07/CTD/*')
files.sort(key=os.path.basename)
#initilize arrays
nsamps=len(files) #number of samples
ss_depth = np.empty((nsamps,500)); ss_depth.fill('NaN')
ss_sal = np.empty((nsamps,500)); ss_sal.fill('NaN')
ss_temp = np.empty((nsamps,500)); ss_temp.fill('NaN')
count = np.zeros((nsamps), dtype=np.int)
lats=np.zeros(nsamps); lons=np.zeros(nsamps); times=[]
i_stn=0
#Loop through files
for filename in files:
    lat,lon,time=read_IOS_file(filename,i_stn)
    i_stn=i_stn+1
times=np.array(times)
```

# Plotting

Plan: For each day, interpolate model and observation to a desired depth. Plot interpolated salinity vs. longitude.

```python
def interpolate_depth(sal,depth_array,depth_new):
    """ interpolates a salinity field given at ds to a desire depth """
    f= interp.interp1d(depth_array,sal)
    sali = f(depth_new)
    return sali
```

```python
def compare_fields(depth,date,smin=10,smax=34,tmin=8,tmax=11,figsize=(20,8)):
    
    fig=plt.figure(figsize=figsize)
    #Isolate model date
    tind = date.day - 21
    time_m=nc_tools.timestamp(gridT,tind)
    
    #Loop through obervations. For ones on date, plot salinity vs longitude for obs and model.
    sal_model=[];sal_obs=[]; lon_plotm=[]; lon_ploto=[]; lat_ploto=[]; lat_plotm=[]
    temp_model=[]; temp_obs=[]
    for n in range(times.shape[0]):
        if times[n].day == date.day:
            #observations data
            obs_plot=interpolate_depth(ss_sal[n,:],ss_depth[n,:],depth)
            sal_obs.append(obs_plot); lon_ploto.append(lons[n]); lat_ploto.append(lats[n])
            obs_plot=interpolate_depth(ss_temp[n,:],ss_depth[n,:],depth)
            temp_obs.append(obs_plot);
            
            # model data
            [j,i]=tidetools.find_closest_model_point(lons[n],lats[n],X,Y,bathy,allow_land=False)
            if not isinstance(j,list): #check in domain
                #salinity
                sal_plot_arr=sal_m[tind,:,j,i]
                #mask to be safe
                mu=sal_plot_arr==0; sal_plot_arr=np.ma.array(sal_plot_arr,mask=mu)
                dz=np.ma.array(depth_m[:],mask=mu);
                #interpolate
                sal_plot=interpolate_depth(sal_plot_arr,dz,depth)
                sal_model.append(sal_plot);lon_plotm.append(lons[n]); lat_plotm.append(lats[n])
                #temperature
                temp_plot_arr=temp_m[tind,:,j,i]
                #mask to be safe
                mu=temp_plot_arr==0; temp_plot_arr=np.ma.array(temp_plot_arr,mask=mu)
                dz=np.ma.array(depth_m[:],mask=mu);
                #interpolate
                temp_plot=interpolate_depth(temp_plot_arr,dz,depth)
                temp_model.append(temp_plot);
           

    #sort so plot is in a line along longitude
    model_sort = [x for y, x in sorted(zip(lon_plotm, sal_model))]
    temp_model_sort = [x for y, x in sorted(zip(lon_plotm, temp_model))]
    lat_model_sort= [x for y, x in sorted(zip(lon_plotm, lat_plotm))]
    lon_plotm.sort()
    obs_sort = [x for y, x in sorted(zip(lon_ploto, sal_obs))]
    temp_obs_sort = [x for y, x in sorted(zip(lon_ploto, temp_obs))]
    lat_obs_sort = [x for y, x in sorted(zip(lon_ploto, lat_ploto))]
    lon_ploto.sort()
    
    #salinty plot
    ax=plt.subplot(2,2,1)
    ax.plot(lon_plotm,model_sort,'bo-',label='Model')
    ax.plot(lon_ploto,obs_sort,'rs--',label='Observation')
    ax.legend(loc=0)
    ax.set_ylabel('Practical Salinity [psu]'); ax.set_xlabel('Longitude')
    ax.set_title('Salinity on {date} at {depth} m'.format(date=date.strftime('%d-%b-%Y'),depth=depth))
    ax.set_ylim([smin,smax])
    ax.set_xlim([-125,-122.3])
    #temperature plot
    ax2 = plt.subplot(2,2,3)
    ax2.plot(lon_plotm,temp_model_sort,'bo-',label='Model')
    ax2.plot(lon_ploto,temp_obs_sort,'rs--',label='Observation')
    ax2.legend(loc=0)
    ax2.set_xlim([-125,-122.3])
    ax2.set_ylim([tmin,tmax])
    ax2.set_ylabel('Temperature (deg C)'); ax.set_xlabel('Longitude')
    ax2.set_title('Temperature on {date} at {depth} m'.format(date=date.strftime('%d-%b-%Y'),depth=depth))
    
    #map
    ax=plt.subplot(1,2,2)
    ax.plot(lon_plotm,lat_model_sort,'bo-',label='Model')
    ax.plot(lon_ploto,lat_obs_sort,'rs-',label='Observation')
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    viz_tools.plot_coastline(ax,grid_B,coords='map')
    
    return fig
```
