```python
def read_IOS_file(filename,i_stn):
    filehandle = open(filename,'rt')
    with filehandle as data:
            headerover = False
            for line in data:
                if headerover:
                    word = line.split()
                    ss_depth[i_stn,count[i_stn]] = float(word[0])
                    ss_sal[i_stn,count[i_stn]] = float(word[5])
                    ss_temp[i_stn,count[i_stn]] = float(word[1])
                    count[i_stn] = count[i_stn] + 1
                elif line[0:14] == "*END OF HEADER":
                    headerover = True
                else:
                    word = line.split()
                    if len(word) > 1:
                        if word[0] == "LATITUDE":
                            lat = float(word[2])+ float(word[3])/60
                            lats[i_stn]=lat
                        if word[0] == "LONGITUDE":
                            lon = -(float(word[2])+ float(word[3])/60)
                            lons[i_stn]=lon
                        if word[0] == "START":
                            string = word[4] +' ' + word[5]
                            time = datetime.datetime.strptime(string,'%Y/%m/%d %H:%M:%S.%f')
                            times.append(time)
    return lat,lon,time
```

```python
files =glob.glob('/ocean/sallen/allen/research/MEOPAR/IOS_2003_CTD/2003-07/CTD/*')
files.sort(key=os.path.basename)
#initilize arrays
nsamps=len(files) #number of samples
ss_depth = np.empty((nsamps,500)); ss_depth.fill('NaN')
ss_sal = np.empty((nsamps,500)); ss_sal.fill('NaN')
ss_temp = np.empty((nsamps,500)); ss_temp.fill('NaN')
count = np.zeros((nsamps), dtype=np.int)
lats=np.zeros(nsamps); lons=np.zeros(nsamps); times=[]
i_stn=0
#Loop through files
for filename in files:
    lat,lon,time=read_IOS_file(filename,i_stn)
    i_stn=i_stn+1
times=np.array(times)
```

```python
def interpolate_depth(sal,depth_array,depth_new):
    """ interpolates a salinity field given at ds to a desire depth """
    f= interp.interp1d(depth_array,sal)
    sali = f(depth_new)
    return sali
```

```python
def compare_salinity(depth,model_ind,smin=10,smax=34,figsize=(20,8)):
    date=datetime.datetime(2003,4,24)
    
    fig,axs=plt.subplots(1,2,figsize=figsize)
    #Isolate model date
    tind = model_ind
    time_m=nc_tools.timestamp(fT,tind)
    
    #Loop through obervations. For ones on date, plot salinity vs longitude for obs and model.
    sal_model={name1: [], name2: []};  lon_plotm={name1: [],name2:[] } ;lat_plotm={name1: [],name2:[]}
    sal_obs=[]; lon_ploto=[]; lat_ploto=[]; 
    for n in range(times.shape[0]):
        if times[n].day == date.day:
            #observations data
            obs_plot=interpolate_depth(ss_sal[n,:],ss_depth[n,:],depth)
            sal_obs.append(obs_plot); lon_ploto.append(lons[n]); lat_ploto.append(lats[n])
            
            # model data
            for name in (name1,name2):
                [j,i]=tidetools.find_closest_model_point(lons[n],lats[n],X,Y,bathy,allow_land=False)
                if not isinstance(j,list): #check in domain
                    sal_plot_arr=sals[name][tind,:,j,i]
                    #mask to be safe
                    mu=sal_plot_arr==0; sal_plot_arr=np.ma.array(sal_plot_arr,mask=mu)
                    dz=np.ma.array(deps[name][:],mask=mu);
                    #interpolate
                    sal_plot=interpolate_depth(sal_plot_arr,dz,depth)
                    sal_model[name].append(sal_plot);
                    lon_plotm[name].append(lons[n]); lat_plotm[name].append(lats[n])
           

    #sort so plot is in a line along longitude
    model_sort={}; lat_model_sort={}
    for name in (name1,name2):
        model_sort[name] = [x for y, x in sorted(zip(lon_plotm[name], sal_model[name]))]
        lat_model_sort[name]= [x for y, x in sorted(zip(lon_plotm[name], lat_plotm[name]))]
        lon_plotm[name].sort()
    obs_sort = [x for y, x in sorted(zip(lon_ploto, sal_obs))]
    lat_obs_sort = [x for y, x in sorted(zip(lon_ploto, lat_ploto))]
    lon_ploto.sort()
    
    #plot
    ax=axs[0]
    for name in (name1,name2):
        ax.plot(lon_plotm[name],model_sort[name],'o-',label=name)
    ax.plot(lon_ploto,obs_sort,'rs-',label='Observation')
    ax.legend(loc=0)
    ax.set_ylabel('Salinity [psu]'); ax.set_xlabel('Longitude')
    ax.set_title('Salinity at model output time {ind} at {depth} m'.format(ind=model_ind,depth=depth))
    ax.set_ylim([smin,smax])
    ax.set_xlim([-125,-122.3])
    
    #map
    ax=axs[1]
    for name in (name1,name2):
        ax.plot(lon_plotm[name],lat_model_sort[name],'-o',label=name)
    ax.plot(lon_ploto,lat_obs_sort,'-o',label='Observation')
    viz_tools.plot_coastline(ax,grid_B,coords='map')
    
    return fig
```
