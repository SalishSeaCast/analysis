Find Tidal Ellipses in CODAR region

```python
# imports
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
import netCDF4 as NC
from scipy.optimize import curve_fit
from salishsea_tools import tidetools
from salishsea_tools import viz_tools
from matplotlib.patches import Ellipse
```

```python
name = '../../myResults/TidesM2phaseEVD10/CODAR_0040.nc'
fn = NC.Dataset(name,'r')
time = fn.variables["time_counter"][:]/3600.  # want hours not seconds
lats = fn.variables["nav_lat"][:]
lons = fn.variables["nav_lon"][:]
v = fn.variables["vomecrty"][:]
u = fn.variables["vozocrtx"][:]
```

```python
print u.shape
thesize = u[0,0,:-1,:-1].shape
print thesize
```

```python
plt.pcolormesh(v[0,0])
plt.colorbar()
```

```python
# frequencies
M2freq = 28.984106 # degrees per hour
M2freq = M2freq*np.pi/180. # radians per hour
K1freq = 15.041069*np.pi/180.
# initialize fit
vM2amp = np.ones(thesize); vM2pha = 180.*np.ones(thesize)
vK1amp = np.ones(thesize); vK1pha = 180.*np.ones(thesize)
uM2amp = np.ones(thesize); uM2pha = 180.*np.ones(thesize)
uK1amp = np.ones(thesize); uK1pha = 180.*np.ones(thesize)
```

```python
# function for fit
def double(x, M2amp, M2pha, K1amp, K1pha):
    return (M2amp*np.cos(M2freq*x-M2pha*np.pi/180.)+
            K1amp*np.cos(K1freq*x-K1pha*np.pi/180.))
```

need to unstagger the velocities, before I calculate the components
so v is north/south and that's the first index
so 2 vus = v[i,j]+v[i-1,j] to put it on the T-grid
2 uus = u[i,j]+u[i,j-1] to put it on the T-grid

```python
ts = 240 # last half of series
id = 15 # depth
iss = range(1,thesize[0]+1)
im1 = range(0,thesize[0])
jss = range(1,thesize[1]+1)
jm1 = range(0,thesize[1])
uaus = 0.5*(u[ts:,id,0:thesize[0],jss]+u[ts:,id,0:thesize[0],jm1])
vaus = 0.5*(v[ts:,id,iss,0:thesize[1]]+v[ts:,id,im1,0:thesize[1]])
print vaus.shape
```

```python
print thesize, np.arange(thesize[0])
for i in np.arange(thesize[0]):
    for j in np.arange(thesize[1]):
        fitted, cov = curve_fit(double,time[ts:],vaus[:,i,j]) 
        if fitted[0] < 0:
            fitted[0] = -fitted[0]
            fitted[1] = fitted[1]+180.
        if fitted[2] < 0:
            fitted[2] = -fitted[2]
            fitted[3] = fitted[3]+180.
        vM2amp[i,j] = fitted[0]
        vM2pha[i,j] = fitted[1]
        vK1amp[i,j] = fitted[2]
        vK1pha[i,j] = fitted[3]
for i in np.arange(thesize[0]):
    for j in np.arange(thesize[1]):
        fitted, cov = curve_fit(double,time[ts:],uaus[:,i,j]) 
        if fitted[0] < 0:
            fitted[0] = -fitted[0]
            fitted[1] = fitted[1]+180.
        if fitted[2] < 0:
            fitted[2] = -fitted[2]
            fitted[3] = fitted[3]+180.
        uM2amp[i,j] = fitted[0]
        uM2pha[i,j] = fitted[1]
        uK1amp[i,j] = fitted[2]
        uK1pha[i,j] = fitted[3]
```

```python
fig = plt.figure(figsize=(2*5.5+1,2*6.4))
plt.subplot(2,2,1)
plt.pcolormesh(vM2amp)
plt.colorbar()
plt.title('v M2 amp')
plt.ylim(0,thesize[0])
plt.xlim(0,thesize[1])
plt.subplot(2,2,2)
plt.pcolormesh(vM2pha)
plt.colorbar()
plt.title('v M2 pha')
plt.ylim(0,thesize[0])
plt.xlim(0,thesize[1])
plt.subplot(2,2,3)
plt.pcolormesh(uM2amp)
plt.colorbar()
plt.title('u M2 amp')
plt.ylim(0,thesize[0])
plt.xlim(0,thesize[1])
plt.subplot(2,2,4)
plt.pcolormesh(uM2pha)
plt.colorbar()
plt.title('u M2 pha')
plt.ylim(0,thesize[0])
plt.xlim(0,thesize[1])
```

```python
CX = uM2amp*np.cos(np.pi*uM2pha/180.)
SX = uM2amp*np.sin(np.pi*uM2pha/180.)
CY = vM2amp*np.cos(np.pi*vM2pha/180.)
SY = vM2amp*np.sin(np.pi*uM2pha/180.)
ap = np.sqrt((CX+SY)**2+(CY-SX)**2)/2.
am = np.sqrt((CX-SY)**2+(CY+SX)**2)/2.
ep = np.arctan2(CX+SY,CY-SX)
em = np.arctan2(CY+SX,CX-SY)
major = ap+am
minor = ap-am
theta = (ep+em)/2.*180./np.pi
print theta.shape, theta[30,10], theta[0,43], theta[0,44], theta[0,45], ap[0,46]
fig=plt.figure(figsize=(8,8))
ax = fig.add_subplot(221)
plt.pcolormesh(ap)
plt.colorbar()
plt.title('ap')
plt.subplot(222)
plt.pcolormesh(am)
plt.colorbar()
plt.title('am')
plt.subplot(223)
plt.pcolormesh(major)
plt.colorbar()
plt.title('major')
plt.subplot(224)
plt.pcolormesh(minor)
plt.colorbar()
plt.title('minor')
        
```
