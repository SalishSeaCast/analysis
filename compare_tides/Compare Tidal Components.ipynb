Lets look at Mike's Data

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def radians(degrees):
    rad = np.pi/180.*degrees
    return rad
```

```python
nstn = 25
K1amp = np.zeros(nstn); Q1amp = np.zeros(nstn); O1amp = np.zeros(nstn); P1amp = np.zeros(nstn)
K1phase = np.zeros(nstn); Q1phase = np.zeros(nstn); O1phase = np.zeros(nstn); P1phase = np.zeros(nstn)
M2amp = np.zeros(nstn); N2amp = np.zeros(nstn); S2amp = np.zeros(nstn); K2amp = np.zeros(nstn)
M2phase = np.zeros(nstn); N2phase = np.zeros(nstn); S2phase = np.zeros(nstn); K2phase = np.zeros(nstn)
filename = 'vigf7_pres_ele_harm_obs2.dat'
fp = open(filename)
for i in range(nstn):
    line = fp.readline()
    lon = line[1:9]
    lat = line[12:19]
    name = line[25:43]
    if i < 20:
        print i, name
    # Diurnals
    line = fp.readline()
    values = line.split(' ')
    diurnals = [float(e) for e in values if (e != '' and e != '\n')]
    K1amp[i] = diurnals[6]
    Q1amp[i] = diurnals[0]/diurnals[6]
    O1amp[i] = diurnals[2]/diurnals[6]
    P1amp[i] = diurnals[4]/diurnals[6]
    K1phase[i] = diurnals[7]
    Q1phase[i] = diurnals[1]-diurnals[7]
    O1phase[i] = diurnals[3]-diurnals[7]
    P1phase[i] = diurnals[5]-diurnals[7]
    # Semi-diurnals
    line = fp.readline()
    values = line.split(' ')
    semis = [float(e) for e in values if (e != '' and e != '\n')]
    M2amp[i] = semis[2]
    N2amp[i] = semis[0]/semis[2]
    S2amp[i] = semis[4]/semis[2]
    K2amp[i] = semis[6]/semis[2]
    M2phase[i] = semis[3]
    N2phase[i] = semis[1]-semis[3]
    if (N2phase[i] > 0.):
        N2phase[i] = N2phase[i]-360.
    S2phase[i] = semis[5]-semis[3]
    if (S2phase[i] < 0.):
        S2phase[i] = S2phase[i]+360.
    K2phase[i] = semis[7]-semis[3]
    if (K2phase[i] < 0.):
        K2phase[i] = K2phase[i]+360.
```

```python
plt.figure(figsize=(10,15))
plt.subplot(3,2,1)
plt.plot(Q1amp,'*', label="Q1")
plt.plot(O1amp,'x', label="O1")
plt.plot(P1amp,'o', label="P1")
plt.title("Amplitude Ratio")
plt.subplot(3,2,2)
plt.plot(Q1phase,'*')
plt.plot(O1phase,'x')
plt.plot(P1phase,'o')
plt.title("Phase Difference (degrees)")
Q1am = np.mean(Q1amp)
Q1pm = np.mean(Q1phase)
Q1aest = K1amp*Q1am
Q1pest = K1phase+Q1pm
O1am = np.mean(O1amp)
O1pm = np.mean(O1phase)
O1aest = K1amp*O1am
O1pest = K1phase+O1pm
P1am = np.mean(P1amp)
P1pm = np.mean(P1phase)
P1aest = K1amp*P1am
P1pest = K1phase+P1pm
plt.subplot(3,2,3)
plt.plot(Q1aest-Q1amp*K1amp,'*')
plt.plot(O1aest-O1amp*K1amp,'x')
plt.plot(P1aest-P1amp*K1amp,'o')
plt.title("Amplitude Error if using Ratio (cm)")
plt.subplot(3,2,4)
plt.plot(Q1pest-(Q1phase+K1phase),'*', label="Q1")
plt.plot(O1pest-(O1phase+K1phase),'x', label="O1")
plt.plot(P1pest-(P1phase+K1phase),'o', label="P1")
plt.legend(loc="upper left")
plt.title("Phase Error if using Difference (degrees)")
CER_Q1 = np.sqrt( (Q1aest*np.cos(radians(Q1pest))
                - (Q1amp*K1amp)*np.cos(radians(Q1phase+K1phase)))**2
                + (Q1aest*np.sin(radians(Q1pest))
                - (Q1amp*K1amp)*np.sin(radians(Q1phase+K1phase)))**2)
CER_O1 = np.sqrt( (O1aest*np.cos(radians(O1pest))
                - (O1amp*K1amp)*np.cos(radians(O1phase+K1phase)))**2
                + (O1aest*np.sin(radians(O1pest))
                - (O1amp*K1amp)*np.sin(radians(O1phase+K1phase)))**2)
CER_P1 = np.sqrt( (P1aest*np.cos(radians(P1pest))
                - (P1amp*K1amp)*np.cos(radians(P1phase+K1phase)))**2
                + (P1aest*np.sin(radians(P1pest))
                - (P1amp*K1amp)*np.sin(radians(P1phase+K1phase)))**2)
plt.subplot(3,2,5)
plt.plot(CER_Q1,'*')
plt.plot(CER_O1,'x')
plt.plot(CER_P1,'o')
plt.title("Complex Error if using Ratio (cm)")
```
