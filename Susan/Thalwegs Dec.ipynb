```python
from __future__ import division

import matplotlib.pyplot as plt
import netCDF4 as nc
import numpy as np

from salishsea_tools import nc_tools

%matplotlib inline
```

```python
fT2002 = nc.Dataset('/results/SalishSea/spin-up/2002/30nov9dec/SalishSea_1d_20021130_20021209_grid_T.nc','r')
sal2002 = fT2002.variables['vosaline'][1,:]
depth = fT2002.variables['deptht'][:]
T_lat = fT2002.variables['nav_lat']
T_lon = fT2002.variables['nav_lon']
m = sal2002 == 0
sal2002 = np.ma.array(sal2002,mask=m)

fT2003 = nc.Dataset('/results/SalishSea/spin-up/27nov6dec/SalishSea_1d_20031127_20031206_grid_T.nc','r')
sal2003 = fT2003.variables['vosaline'][4,:]
m = sal2003 == 0
sal2003 = np.ma.array(sal2003,mask=m)

fT2014 = nc.Dataset('/results/SalishSea/nowcast/01dec14/SalishSea_1d_20141201_20141201_grid_T.nc','r')
sal2014 = fT2014.variables['vosaline'][0,:]
m = sal2014 == 0
sal2014 = np.ma.array(sal2014,mask=m)

fT2015 = nc.Dataset('/results/SalishSea/nowcast/01dec15/SalishSea_1d_20151201_20151201_grid_T.nc','r')
sal2015 = fT2015.variables['vosaline'][0,:]
m = sal2015 == 0
sal2015 = np.ma.array(sal2015,mask=m)
```

```python
d=40

smin=26; smax=33; dels=-0.4
smin = 24.; smax = 33.5; dels=-0.1

lines = np.loadtxt('../../Tools/bathymetry/thalweg_working.txt', delimiter=" ", unpack=False)
lines = lines.astype(int)

thalweg_lon = T_lon[lines[:,0],lines[:,1]]
thalweg_lat = T_lat[lines[:,0],lines[:,1]]

ds=np.arange(0,lines.shape[0],1);

vs= [29, 30, 31]
XX,ZZ = np.meshgrid(ds,-depth[:d])

salT=sal2002[:,lines[:,0],lines[:,1]]

plt.figure(figsize=(25,10))
plt.pcolormesh(XX,ZZ,salT[:d],vmin=smin,vmax=smax,cmap='spectral')
plt.colorbar()
CS=plt.contour(XX,ZZ,salT[:d],vs, colors='black')
plt.clabel(CS, fontsize=9, inline=1)
plt.title('2002')
```
