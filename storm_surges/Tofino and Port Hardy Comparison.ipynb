This notebook compares the hourly sea surface height anomaly at Tofino and Port Hardy. We do not have a surge prediction tool at Port Hardy so this comparison will guide our decisions regarding the northern boundary condition on sea surface height. 

```python
%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np

import arrow
from salishsea_tools import stormtools
```

First, prepare to load in the ssh forcing data. Let's make a function for that.

```python
def compare_ssh(date):
    #plot ssh anomaly for Port Hardy and Tofino for the month given in date
    #date should be in the format '01-Dec-2006'
    #Note that we only have 200 data at the moment.
    
    #directory paths for data
    ssh_Tofino = '/data/nsoontie/MEOPAR/NEMO-forcing/open_boundaries/west/ssh/ssh'

    ssh_PH = '/data/nsoontie/MEOPAR/NEMO-forcing/open_boundaries/north/ssh/sshNorth'

    ssh={}; time={};

    arr_start =arrow.Arrow.strptime(date,'%d-%b-%Y')
    yr=arr_start.year
    m=arr_start.month
    if m<10:
        m='0' + str(m)
    string = '_y'+ str(yr) + 'm' + str(m) +'.nc'
    
    #get the data
    fT = NC.Dataset(ssh_Tofino + string);
    ssh['Tofino']=fT.variables['sossheig'];
    tss= fT.variables['time_counter'][:];
    l = tss.shape[0]; t=np.linspace(0,l-1,l) #time array
    time['Tofino']=stormtools.convert_date_hours(t,date);

    fPH=NC.Dataset(ssh_PH + string);
    ssh['Port Hardy']=fPH.variables['sossheig'];
    tss= fPH.variables['time_counter'][:];
    l = tss.shape[0]; t=np.linspace(0,l-1,l) #time array
    time['Port Hardy']=stormtools.convert_date_hours(t,date);
    
    #plot the data
    plt.figure(figsize=(18,5))
    ax1=pylab.subplot(1,2,1)
    for key in ssh:
        pylab.plot(time[key],ssh[key][:,0,0],label=key)
    ax1.set_ylim([-1,1])
    pylab.xlabel('Time')
    pylab.ylabel('SSH anomaly (m)')
    pylab.grid()
    pylab.legend(loc=0)
    pylab.title('SSH Anomaly for ' +date)

    
    ax2=pylab.subplot(1,2,2)
    diff =ssh['Tofino'][:,0,0]-ssh['Port Hardy'][:,0,0] 
    mean = np.mean(diff)
    ma = np.max(diff)
    mi = np.min(diff)
    pylab.plot(time[key],diff,'r',label='difference')
    pylab.plot(time[key],np.mean(diff)*np.ones(np.size(time[key])),'--k', label='mean of difference')
    ax2.set_ylim([-.5,.5])
    pylab.xlabel('Time')
    pylab.ylabel('Difference in anomaly (m)')
    pylab.title('Tofino-Port Hardy')
    pylab.grid()
   
    print 'Mean difference ' + str(mean)
    print 'Max: ' + str(ma)
    print 'Min: ' +str(mi)
```

# Summary

1. Both signals follow the same general trend.

2. On days of known storm surges, the Tofino signal is typically larger than the Port Hardy signal. Somtimes the Tofino signal is 50cm larger. The timing of the surge events matches between the two stations. 

3. The Port Hardy signal contains fairly large daily oscillations, especially in the summer months. This seems funny. I should double check on how the anomaly was calculated. 

4. This is only one year. There is a clear seasonal pattern in the anomalies; large anomalies in winter months, not so much in summer months. 2006 was a weak El Nino year according to this website: http://ggweather.com/enso/oni.htm
