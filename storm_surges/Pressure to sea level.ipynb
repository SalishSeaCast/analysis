# Height above Sea Level

Give the temperature $T_1$ and pressure $p_1$ of an air parcel, we can approximate its height $z_1$ above sea level using the following formula:

$z_1 = \frac{T_1}{\gamma}\left( \left(\frac{p_1}{p_0}\right)^{-\frac{\gamma R}{g} } - 1 \right) $

where $g$ is the acceleration due to gravity, $R$ is the ideal gas constant, $p_0$ is the sea level pressure, and $\gamma$ is the temperature lapse rate of the atmosphere.

To arrive at this formula, we have made a few assumptions
1. The temperature of the atmosphere decreases with height at a constant rate $\frac{dT}{dz} = -\gamma$.
2. The atmosphere is an ideal gas, $p=\rho RT$, where $\rho$ is the density.
3. The atmosphere is in hydrostatic equilibrium $\frac{dp}{dz} = -\rho g$

We will take an average slp over an ocean grid point in CGRF. This is hopefully close to 101kPa. Note: does it make more sense to use 101kPa? NEMO calculates inverse barometer based on difference from 101 kPa. 

We can apply this formula to find the altitude of the CGRF points

```python
def altitude(P,T,p0):
    R = 287 #ideal gas constant
    g = 9.81 #gravity
    gam = 0.0098 #lapse rate(deg/m)
    
    z = T/gam*((P/p0)**(-gam*R/g) -1 )
    
    return z
```

We will average the pressure and temperature fields over time to come up with an average height of each grid cell.

```python
pavg = np.mean(press,0)
tavg = np.mean(temp,0)

z = altitude(pavg,tavg,p0)

fig, ax = plt.subplots(1, 1, figsize=(10,10))
mesh=ax.pcolormesh(lonCGRF[xs:xe,ys:ye],latCGRF[xs:xe,ys:ye],z)
viz_tools.plot_coastline(ax,fB,coords='map')
cbar=fig.colorbar(mesh,ax=ax)
ax.set_xlim([-126.5,-121.5])
ax.set_ylim([47,51.2])
```

# Sea Level Pressure

Assuming the altitude $z_1$ at each grid cell is computed as above, we can determine the sea level pressure of each grid point at each time using the following formula:

$ p_s = p_1\left(\gamma \frac{z_1}{T_1} +1 \right)^{\frac{g}{\gamma R}} $

```python
def slp(Z,P,T):
    R = 287 #ideal gas constant
    g = 9.81 #gravity
    gam = 0.0098 #lapse rate(deg/m)
    
    ps = P*(gam*(Z/T) +1)**(g/gam/R)
    
    return ps
```

```python
vmin=95000; vmax=105000
vs=np.arange(vmin,vmax,20)
(fig,axs )= plt.subplots(4,6,figsize=(30,30))
ts=np.arange(24)
for (ax,t) in zip(axs.flat,ts):
    ps = slp(z,press[t,:,:],temp[t,:,:])
    mesh=ax.pcolormesh(lonCGRF[xs:xe,ys:ye],latCGRF[xs:xe,ys:ye],ps,vmin=vmin, vmax= vmax)
    viz_tools.plot_coastline(ax,fB,coords='map')
    ax.set_title(t)
    cbar=fig.colorbar(mesh,ax=ax)
    ax.set_xlim([-126.5,-121.5])
    ax.set_ylim([47,51.2])
```

```python
#prepare for getting EC Pressure at YVR
def get_pressure_obs(station, start_day, end_day):
    station_ids = {
        'Sandheads': 6831,
        'YVR': 889,
        'PointAtkinson': 844,
        'Victoria': 10944,
        'CampbellRiver': 145,
        'PatriciaBay': 11007
    }
    wind_spd, pressure = [], []
    url = 'http://climate.weather.gc.ca/climateData/bulkdata_e.html'
    query = {
        'timeframe': 1,
        'stationID': station_ids[station],
        'format': 'xml',
        'Year': 2006,
        'Month': 2,
        'Day': 1,
    }
    response = requests.get(url, params=query)
    tree = ElementTree.parse(cStringIO.StringIO(response.content))
    root = tree.getroot()
    raw_data = root.findall('stationdata')
    for record in raw_data:
        day = int(record.get('day'))
        hour = int(record.get('hour'))
        year = int(record.get('year'))
        month = int(record.get('month'))
        selectors = (
            (day == start_day - 1 and hour >= 16)
            or
            (day >= start_day and day < end_day)
            or
            (day == end_day and hour < 16)
        )
        if selectors:
            try:
                wind_spd.append(float(record.find('windspd').text))
            except TypeError:
                wind_spd.append(0)
            if station == 'YVR':
                try:
                    pressure.append(float(record.find('stnpress').text) )
                except ValueError:
                    pressure.append(0)
    wind_spd= np.array(wind_spd) * 1000 / 3600
    pressure = np.array(pressure)
    
    return wind_spd, pressure
```

```python
#YVR pressure - obs
[wind,press_o] = get_pressure_obs('YVR', 4, 4) 

fig,axs=plt.subplots(1,2,figsize=(15,3))
ax=axs[0]
ax.plot(press_o*1000,label='obs')
ax.plot(pss[:,point[0],point[1]],label='corrected CGRF')
ax.plot(press[:,point[0],point[1]],label='uncorrected CGRF')
ax.legend(loc=0)
ax.set_xlabel('time on Feb 4')

ax=axs[1]
ax.plot(press_o*1000-pss[:,point[0],point[1]])
ax.set_title('Difference')
ax.set_xlabel('time on Feb 4')
```
