This notebook will examine the storm surges under different bottom friction parametrizations. 

Cases:

1. control - bfrib2 = 5e-3, bfeb2 = 2.5e-3, rhn_shlat=0.5
2. noeb - bfrib2 = 5e-3, bfeb2 = 0
3. bfrib3e3 - bfrib2 = 3e-3, bfeb2=0
4. partial - bfrib2 = 5e-3, bfeb2 = 2.5e-3, rn_shlat = 0.1

```python
%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np

import arrow
from salishsea_tools import stormtools
from salishsea_tools import tidetools
```

```python
path = '/data/nsoontie/MEOPAR/SalishSea/results/storm-surges/bottom_friction/'

run1= 'control';
run2='noeb';
run3='bfri3e3'; 
run4='partial'
runs = {run1, run2, run3,run4}

fUs={}; fVs={}; fTs={};

for key in runs:
    fUs[key] = NC.Dataset(path + key +'/SalishSea_4h_20061213_20061215_grid_U.nc','r');
    fVs[key] = NC.Dataset(path + key +'/SalishSea_4h_20061213_20061215_grid_V.nc','r');
    fTs[key] = NC.Dataset(path + key +'/SalishSea_4h_20061213_20061215_grid_T.nc','r');
```

```python
# define plot parameters
#time, depth, colormap etc
timestamp =15#model output time 
times = fVs[key].variables['time_counter'] #array of output times
time = times[timestamp]  #physical time in seconds
depthlevel = 0  #model level to plot
depths = fVs[key].variables['depthv'] #array of depths
depth = depths[depthlevel]  #physical depth in meters
cmap = plt.get_cmap('jet') 
u_lat = fUs[key].variables['nav_lat']
u_lon = fUs[key].variables['nav_lon']
T_lat = fTs[key].variables['nav_lat']
T_lon = fTs[key].variables['nav_lon']
v_lat = fVs[key].variables['nav_lat']
v_lon = fVs[key].variables['nav_lon']
```

```python
Us={}; Vs={}; Es ={}; Ss={}; Ts={};

for key in runs:
    [Us[key], Vs[key], Es[key], Ss[key], Ts[key]] = stormtools.get_variables(fUs[key],fVs[key],fTs[key],timestamp,depthlevel)
```

```python

start='13-Dec-2006'
stations = {'PointAtkinson': 7795, 'Victoria': 7120, 'PatriciaBay': 7277, 'CampbellRiver': 8074, 'Plume': 1}
datums = {'PointAtkinson': 3.09, 'Victoria': 1.881, 'PatriciaBay': 2.256, 'CampbellRiver': 2.916, 'Plume': 1}
thalwegs = ['Thalweg1','Thalweg2', 'Thalweg3', 'Thalweg4', 'Thalweg5', 'Thalweg6']
allstations = {'PointAtkinson': 7795, 'Victoria': 7120, 'PatriciaBay': 7277, 'CampbellRiver': 8074, 'Plume': 1,
               'Thalweg1': [], 'Thalweg2': [], 'Thalweg3': [], 'Thalweg4': [], 'Thalweg5': [], 'Thalweg6': []} 
#datums from t_xtide
run_stations={}
us={}; vs={}; lats={}; lons={}; tmps={}; sals={}; sshs={}; ts={};

for run in runs:
    for key in allstations:
        string = path + run + '/1h_' + key + '.nc'
        run_stations[key] = NC.Dataset(string,'r');
        t=run_stations[key].variables['time_counter']
        ts[run]=stormtools.convert_date_seconds(t,start)
    [us[run], vs[run], lats[run], lons[run], tmps[run], sals[run], sshs[run]] = stormtools.combine_data(run_stations)
    run_stations={};
```

Compare between the control case and turning off parametrized bottom friction.

```python
start='13-Dec-2006'
end='15-Dec-2006'
emin=-3
emax=3
plt.figure(figsize=(20,12))

obs = {'PointAtkinson': 7795, 'Victoria': 7120, 'PatriciaBay': 7277, 'CampbellRiver': 8074}
datums = {'PointAtkinson': 3.09, 'Victoria': 1.881, 'PatriciaBay': 2.256, 'CampbellRiver': 2.916}



i=1;
for key in obs:
    for run in {run1,run2}:
        ax=pylab.subplot(4,2,2*i-1)
        pylab.plot(ts[run],sshs[run][key][:,0,0],label=run)
        pylab.title(key)
        ax.set_ylim([emin,emax])
        pylab.legend(loc=0)
    pylab.grid()    
        
    ax=pylab.subplot(4,2,2*i)
    pylab.plot(ts[run],sshs[run1][key][:,0,0]-sshs[run2][key][:,0,0],label='difference')
    pylab.title(key)
    ax.set_ylim([-.1,.1])
    pylab.grid()
    i=i+1   
    
pylab.legend( loc=0)
print 'Sea Surface Height from model'
print run1 + '-' + run2 + ' in difference'
```

Storms look identical. Next, check how close the fields are.

```python
print 'Max differences: U,V,E,T,S'
print np.max(Us[run1]-Us[run2])
print np.max(Vs[run1]-Vs[run2])
print np.max(Es[run1]-Es[run2])
print np.max(Ts[run1]-Ts[run2])
print np.max(Ss[run1]-Ss[run2])

print 'Mindifferences: U,V,E,T,S'
print np.min(Us[run1]-Us[run2])
print np.min(Vs[run1]-Vs[run2])
print np.min(Es[run1]-Es[run2])
print np.min(Ts[run1]-Ts[run2])
print np.min(Ss[run1]-Ss[run2])
```

Not much of a difference when removing the tidal friction parametrization, which we aren't supposed to need if forcing tides explicitly.

Next, reduce the friction parameter and examine.

```python
start='13-Dec-2006'
end='15-Dec-2006'
emin=-3
emax=3
plt.figure(figsize=(20,12))

obs = {'PointAtkinson': 7795, 'Victoria': 7120, 'PatriciaBay': 7277, 'CampbellRiver': 8074}
datums = {'PointAtkinson': 3.09, 'Victoria': 1.881, 'PatriciaBay': 2.256, 'CampbellRiver': 2.916}



i=1;
for key in obs:
    for run in {run1,run3}:
        ax=pylab.subplot(4,2,2*i-1)
        pylab.plot(ts[run],sshs[run][key][:,0,0],label=run)
        pylab.title(key)
        ax.set_ylim([emin,emax])
        pylab.legend(loc=0)
    pylab.grid()    
        
    ax=pylab.subplot(4,2,2*i)
    pylab.plot(ts[run],sshs[run1][key][:,0,0]-sshs[run3][key][:,0,0],label='difference')
    pylab.title(key)
    ax.set_ylim([-.1,.1])
    pylab.grid()
    i=i+1   
    
pylab.legend( loc=0)
print 'Sea Surface Height from model'
print run1 + '-' + run3 + ' in difference'
```

```python
print 'Max differences: U,V,E,T,S'
print np.max(Us[run1]-Us[run3])
print np.max(Vs[run1]-Vs[run3])
print np.max(Es[run1]-Es[run3])
print np.max(Ts[run1]-Ts[run3])
print np.max(Ss[run1]-Ss[run3])

print 'Min differences: U,V,E,T,S'
print np.min(Us[run1]-Us[run3])
print np.min(Vs[run1]-Vs[run3])
print np.min(Es[run1]-Es[run3])
print np.min(Ts[run1]-Ts[run3])
print np.min(Ss[run1]-Ss[run3])
```

Up to 5cm differences (lower) in SSH when reducing the bottom friction parameter. I wonder how this would affect the tides. From the difference plot, the lower bottom friction place doesn't always mean higher or lower ssh. So I wonder how this will affect the tides. 

I would expect the currents at least to be stronger in the lower bottom friction case. 

```python
start='13-Dec-2006'
end='15-Dec-2006'
emin=-3
emax=3
plt.figure(figsize=(20,12))

obs = {'PointAtkinson': 7795, 'Victoria': 7120, 'PatriciaBay': 7277, 'CampbellRiver': 8074}
datums = {'PointAtkinson': 3.09, 'Victoria': 1.881, 'PatriciaBay': 2.256, 'CampbellRiver': 2.916}



i=1;
for key in obs:
    for run in {run1,run4}:
        ax=pylab.subplot(4,2,2*i-1)
        pylab.plot(ts[run],sshs[run][key][:,0,0],label=run)
        pylab.title(key)
        ax.set_ylim([emin,emax])
        pylab.legend(loc=0)
    pylab.grid()    
        
    ax=pylab.subplot(4,2,2*i)
    pylab.plot(ts[run],sshs[run1][key][:,0,0]-sshs[run4][key][:,0,0],label='difference')
    pylab.title(key)
    ax.set_ylim([-.1,.1])
    pylab.grid()
    i=i+1   
    
pylab.legend( loc=0)
print 'Sea Surface Height from model'
print run1 + '-' + run4 + ' in difference'
```

Up to 10cm differences in Campbell River. The case with more friction (high rn_shlat) usually has a higher ssh. I'm not sure if this will be usefull for our tides. But, the fact that Campbell River is affected most by the change could indicate that we are too constricted in that region.
