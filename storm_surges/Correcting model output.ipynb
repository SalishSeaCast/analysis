All of the model data is now loaded. Now let's define a function for loading the tidal predictions.

```python
def dateParserMeasured2(s):
    #convert the string to a datetime object
    unaware = datetime.datetime.strptime(s, "%d-%b-%Y %H:%M:%S ")
    #add in the local time zone (Canada/Pacific)
    aware = unaware.replace(tzinfo=pytz.timezone('Canada/Pacific'))
    #convert to UTC
    return aware.astimezone(pytz.timezone('UTC'))

def dateParserMeasured(s):
    #convert the string to a datetime object
    unaware = datetime.datetime.strptime(s, "%Y/%m/%d %H:%M")
    #add in the local time zone (Canada/Pacific)
    aware = unaware.replace(tzinfo=pytz.timezone('Canada/Pacific'))
    #convert to UTC
    return aware.astimezone(pytz.timezone('UTC'))
```

```python
def load_tidal_predictions(start,end,location):
    filename='/data/nsoontie/MEOPAR/analysis/storm_surges/data/'+location+'_t_tide_compare8_' +start+'_'+end+'.csv'
    
    #read msl
    import csv
    line_number = 1     
    with open(filename, 'rb') as f:
        mycsv = csv.reader(f); mycsv = list(mycsv)
        msl = mycsv[line_number][1]
        msl=float(msl)
    
    ttide = pd.read_csv(filename,skiprows=3,parse_dates=[0],date_parser=dateParserMeasured2)
    ttide = ttide.rename(columns={'Time_Local ': 'time', ' pred_8 ': 'pred_8', ' pred_all ': 'pred_all'})
    
    return ttide, msl
```

```python
def load_observations(start,end,location):
    statID_PA = stations[location]
    filename = 'wlev_' +str(statID_PA) + '_' + start +'_' +end +'.csv'
    tidetools.get_dfo_wlev(statID_PA,start,end)
    wlev_meas = pd.read_csv(filename,skiprows=7,parse_dates=[0],date_parser=dateParserMeasured)
    wlev_meas = wlev_meas.rename(columns={'Obs_date': 'time', 'SLEV(metres)': 'slev'})
    
    return wlev_meas
```

```python
start='31-Dec-2005'; end='02-Jan-2007'
wlev_meas={}; ttide={}; msl={}

for key in stations:
    location=key
    wlev_meas[key] = load_observations(start,end,location)
    [ttide[key], msl[key]] = load_tidal_predictions(start,end,location)
```

```python
def compare_tidal_predictions(ttide, start,end,location):

    fig = plt.figure(figsize=(20,10))
    ax=pylab.subplot(1,1,1)
    import matplotlib.dates as mdates
    ax.fmt_xdata = mdates.DateFormatter('%Y-%m-%d')
    fig.autofmt_xdate()

    difference = ttide.pred_all-ttide.pred_8 # difference between the two

    pylab.plot(ttide.time, ttide.pred_8, label='ttide with 8',lw=2)
    pylab.plot(ttide.time, ttide.pred_all, label='ttide with all',lw=2)
    pylab.plot(ttide.time, difference, label="Difference")

    ax.set_xlim([start,end])
    ax.set_ylim([-3,3])
    pylab.legend(loc=0)
    pylab.grid(True,lw=1)
    pylab.title(location)
```

```python
def observed_anomaly(ttide,wlev_meas):
    ssanomaly = np.zeros(len(wlev_meas.time))
    for i in np.arange(0,len(wlev_meas.time)):
    #check that there is a corresponding time 
    #if any(wlev_pred.time == wlev_meas.time[i]):
        ssanomaly[i] =(wlev_meas.slev[i] - (ttide.pred_all[ttide.time==wlev_meas.time[i]]+msl[key]))
        if not(ssanomaly[i]):
            ssanomaly[i]=float('Nan')
        
    return ssanomaly
```

```python
def modelled_anomaly(sshs,location):
    anom=sshs['all_forcing'][location][:,0,0]-sshs['tidesonly'][location][:,0,0]
    return anom
```

```python
def compare_anomaly(sshs,t_model,ttide,wlev_meas,location,start,end):
    anom_obs=observed_anomaly(ttide[location],wlev_meas[location])
    anom_model=modelled_anomaly(sshs,location)
    
    fig = plt.figure(figsize=(20,10))
    ax=pylab.subplot(1,1,1)
    import matplotlib.dates as mdates
    ax.fmt_xdata = mdates.DateFormatter('%Y-%m-%d')
    fig.autofmt_xdate()
    
    difference = ttide[location].pred_all-ttide[location].pred_8 # difference between the two

    pylab.plot(wlev_meas[location].time, anom_obs, 'k', label='Observed anomaly',lw=2)
    pylab.plot(t_model, anom_model, label='Modelled abomaly',lw=2)

    ax.set_xlim([start,end])
    ax.set_ylim([-1.5,1.5])
    pylab.legend(loc=0)
    pylab.grid(True,lw=1)
    pylab.title(location)
```

```python
def correct_model(ssh,ttide,sdt,edt):
    
    #find index of ttide.time at start and end
    inds = ttide.time[ttide.time==sdt].index[0]
    inde = ttide.time[ttide.time==edt].index[0]
    
    difference = ttide.pred_all-ttide.pred_8
    corr = difference[inds:inde]
    
    corr_model = ssh+corr
    return corr_model
```

```python
def compare_ssh_corrected(ssh,t_model,ttide,wlev_meas,msl,start,end,location):
    ssh_corr=correct_model(ssh['all_forcing'][location][:,0,0],ttide[location],start,end)
    
    fig = plt.figure(figsize=(20,10))
    ax=pylab.subplot(1,1,1)
    import matplotlib.dates as mdates
    ax.fmt_xdata = mdates.DateFormatter('%Y-%m-%d')
    fig.autofmt_xdate()

    pylab.plot(wlev_meas[location].time, wlev_meas[location].slev-msl[location], 'k', label='Observations',lw=2)
    pylab.plot(t_model, ssh_corr, 'b--',label='Corrected model',lw=2)

    ax.set_xlim([start,end])
    ax.set_ylim([-3,3])
    pylab.legend(loc=0)
    pylab.grid(True,lw=1)
    pylab.title(location)
```

```python

for key in stations:
    location=key
    compare_ssh_corrected(sshs,ts['all_forcing'],ttide,wlev_meas,msl,sdt,edt,location)
```
