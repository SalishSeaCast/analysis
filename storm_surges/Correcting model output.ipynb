```python
%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np

import arrow
import datetime
import pandas as pd
import pytz
from salishsea_tools import tidetools
from salishsea_tools import stormtools
from salishsea_tools import nc_tools
```

```python
path = '/data/nsoontie/MEOPAR/SalishSea/results/storm-surges/tide_fix/feb2006/'


runs = {'all_forcing','tidesonly', 'surgeonly'}

fUs={}; fVs={}; fTs={};

for key in runs:
    fUs[key] = NC.Dataset(path + key +'/SalishSea_4h_20060201_20060207_grid_U.nc','r');
    fVs[key] = NC.Dataset(path + key +'/SalishSea_4h_20060201_20060207_grid_V.nc','r');
    fTs[key] = NC.Dataset(path + key +'/SalishSea_4h_20060201_20060207_grid_T.nc','r');
```

```python
# define plot parameters
#time, depth, colormap etc
timestamp =41#model output time 
times = fVs[key].variables['time_counter'] #array of output times
time = times[timestamp]  #physical time in seconds
depthlevel = 0  #model level to plot
depths = fVs[key].variables['depthv'] #array of depths
depth = depths[depthlevel]  #physical depth in meters
cmap = plt.get_cmap('jet') 
u_lat = fUs[key].variables['nav_lat']
u_lon = fUs[key].variables['nav_lon']
T_lat = fTs[key].variables['nav_lat']
T_lon = fTs[key].variables['nav_lon']
v_lat = fVs[key].variables['nav_lat']
v_lon = fVs[key].variables['nav_lon']
```

```python
Us={}; Vs={}; Es ={}; Ss={}; Ts={};

for key in runs:
    [Us[key], Vs[key], Es[key], Ss[key], 
     Ts[key]] = stormtools.get_variables(fUs[key],fVs[key],fTs[key],timestamp,depthlevel)
```

```python

stations = {'PointAtkinson': 7795, 'Victoria': 7120, 'PatriciaBay': 7277, 'CampbellRiver': 8074}
datums = {'PointAtkinson': 3.09, 'Victoria': 1.881, 'PatriciaBay': 2.256, 'CampbellRiver': 2.916, 
          'CrescentBeach': 2.44, 'WhiteRock': 2.85, 'BoundaryBay': 1 }
thalwegs = ['Thalweg1','Thalweg2', 'Thalweg3', 'Thalweg4', 'Thalweg5', 'Thalweg6']
allstations = {'PointAtkinson': 7795, 'Victoria': 7120, 'PatriciaBay': 7277, 'CampbellRiver': 8074,
               'CrescentBeach': 2.44, 'WhiteRock': 2.85 , 'BoundaryBay': 1,
               'Thalweg1': [], 'Thalweg2': [], 'Thalweg3': [], 'Thalweg4': [], 'Thalweg5': [], 'Thalweg6': [],
               'Plume': 1} 
#datums from t_xtide
run_stations={}
us={}; vs={}; lats={}; lons={}; tmps={}; sals={}; sshs={}; ts={};

for run in runs:
    for key in allstations:
        string = path + run + '/1h_' + key + '.nc'
        run_stations[key] = NC.Dataset(string,'r');
        tim = run_stations[key].variables['time_counter']
        t_count=np.arange(0, tim.shape[0])
        t=nc_tools.timestamp(run_stations[key],t_count)
        tlist=[]
        for a in t:
           tlist.append(a.datetime) 
        ts[run]=tlist
    [us[run], vs[run], lats[run], lons[run], tmps[run], sals[run], sshs[run]] = stormtools.combine_data(run_stations)
    run_stations={};
```

Load up the observations and predictions now. Makes use of some storm tools

```python
start='31-Dec-2005'; end='02-Jan-2007'
wlev_meas={}; ttide={}; msl={}


for key in stations:
    location=key
    #filename for predictions
    filename='/data/nsoontie/MEOPAR/analysis/storm_surges/data/'+location+'_t_tide_compare8_' +start+'_'+end+'.csv'
    [ttide[key], msl[key]] = stormtools.load_tidal_predictions(filename)
    wlev_meas[key] = stormtools.load_observations(start,end,location)
```

```python
def compare_tidal_predictions(ttide, sdt,edt,location):
    # a function for plotting predictions. sdt and edt are datetime objects that define the x axis range
    fig = plt.figure(figsize=(20,10))
    ax=pylab.subplot(1,1,1)
    import matplotlib.dates as mdates
    ax.fmt_xdata = mdates.DateFormatter('%Y-%m-%d')
    fig.autofmt_xdate()

    difference = ttide.pred_all-ttide.pred_8 # difference between the two

    pylab.plot(ttide.time, ttide.pred_8, label='ttide with 8',lw=2)
    pylab.plot(ttide.time, ttide.pred_all, label='ttide with all',lw=2)
    pylab.plot(ttide.time, difference, label="Difference")

    ax.set_xlim([sdt,edt])
    ax.set_ylim([-3,3])
    pylab.legend(loc=0)
    pylab.grid(True,lw=1)
    pylab.title(location)
```

```python
def compare_anomaly(sshs,t_model,ttide,wlev_meas,location,sdt,edt):
    anom_obs=stormtools.observed_anomaly(ttide[location],wlev_meas[location],msl[location])
    anom_model=stormtools.modelled_anomaly(sshs,location)
    
    fig = plt.figure(figsize=(20,10))
    ax=pylab.subplot(1,1,1)
    import matplotlib.dates as mdates
    ax.fmt_xdata = mdates.DateFormatter('%Y-%m-%d')
    fig.autofmt_xdate()
    
    difference = ttide[location].pred_all-ttide[location].pred_8 # difference between the two

    pylab.plot(wlev_meas[location].time, anom_obs, 'k', label='Observed anomaly',lw=2)
    pylab.plot(t_model, anom_model, label='Modelled anomaly',lw=2)
    pylab.plot(t_model, sshs['surgeonly'][location][:,0,0], 'r',label='Modelled surge',lw=2)

    ax.set_xlim([sdt,edt])
    ax.set_ylim([-1.5,1.5])
    pylab.legend(loc=0)
    pylab.grid(True,lw=1)
    pylab.title(location)
```

```python
def compare_ssh_corrected(ssh,t_model,ttide,wlev_meas,msl,sdt,edt,location):
    ssh_corr=stormtools.correct_model(ssh['all_forcing'][location][:,0,0],ttide[location],sdt,edt)
    surgetide=stormtools.surge_tide(ssh['surgeonly'][location][:,0,0],ttide[location],sdt,edt)
    
    fig = plt.figure(figsize=(20,10))
    ax=pylab.subplot(1,1,1)
    import matplotlib.dates as mdates
    ax.fmt_xdata = mdates.DateFormatter('%Y-%m-%d')
    fig.autofmt_xdate()

    pylab.plot(wlev_meas[location].time, wlev_meas[location].slev-msl[location], 'k', label='Observations',lw=2)
    pylab.plot(t_model, ssh_corr, 'b--',label='Corrected model',lw=2)
    pylab.plot(t_model, surgetide, 'r--',label='Modelled surge +tide',lw=2)

    ax.set_xlim([sdt,edt])
    ax.set_ylim([-3,3])
    pylab.legend(loc=0)
    pylab.grid(True,lw=1)
    pylab.title(location)
```
