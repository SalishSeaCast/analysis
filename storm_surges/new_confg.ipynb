# Statistics

Measuring errors. Do we see an improvement with the new tides and atmospheric pressure files? It is difficult to measure determine this by looking at a picture. 

We will evaluate a statistical measure used in Bernier and Thompson, JGR, 2006 to indicate accuracy of surge predictions.

$\gamma^2 = \frac{Var(\eta_{obs}-\eta_{mod})}{Var(\eta_{obs})} $

This compares the variance in the model error to the variance in observations. Smaller values indicate a a more accurate simulation.

Need a way to caclulate this statistic.
1. grab measurements/model over time period of interest (avoid first 24hours of simulation due to spin up)
2. average measurements over an hour so that time aligns with model
3. Calculate above measure

```python
def rmse(diff):
    return np.sqrt(np.mean(diff**2))
```

```python
def get_slab(ssh, time, sdt, edt):
    #returns an array of water level measurements bewtween start date sdt and end date edt
    inds = np.where(time==sdt)[0]
    inde = np.where(time==edt)[0]
    sub_data=np.array(ssh[inds:inde+1])
    sub_time = np.array(time[inds:inde+1])
    
    return sub_data, sub_time
```

```python
def rebase_meas(wlev,time):
    #rebases the measurements so that they are given on the half hour instead of hour.
    rebase = 0.5*(wlev[1:]+wlev[:-1])
    rebase_t=[]
    for k in range(time.shape[0]):
        rebase_t.append(time[k].replace(minute=30))

    rebase_t=np.array(rebase_t)
    rebase_t=rebase_t[0:-1]
    return rebase, rebase_t
```

```python
def calculate_stats(ssh_model, time_model, ssh_meas, time_meas, sdt, edt):
    #get the slab of model data
    slab_model, slab_tm = get_slab(ssh_model, time_model, sdt.replace(minute=30), edt.replace(minute=30))
    slab_model = slab_model[:-1]
    slab_tm = slab_tm[:-1]
    #get slab of measurements
    slab_meas,slab_tmeas=get_slab(ssh_meas,time_meas,sdt,edt)
    #rebase measuremnts
    rbase_meas, rbase_tmeas=rebase_meas(slab_meas,slab_tmeas)

    #calculate statisitc
    gam2 = np.var(rbase_meas-slab_model)/np.var(rbase_meas)
    me = np.mean(rbase_meas-slab_model)
    rms= rmse(rbase_meas-slab_model)
    
    return gam2, me, rms
```
