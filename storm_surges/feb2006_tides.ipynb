This notebook analyses a Feb 2006 storm surge with the corrected tides. K1 is phase/amp corrected, M2 is phase shifted, lowered bottom firction to 1e-3. M2 is set to match observations at Point Atkinson. Cases:

1. all_forcing - all tidal forcing and ssh forcing and meteorolgical forcing
2. tidesonly - no meteorological forcing and no ssh forcing. Only tidal forcing.
3. ssh_only - no meerological
4. weather_only - no ssh forcing
5. no_pressure - no ssh and no inverse barometer pressure
6. runoff_test - tidesonly paramters with ln_rnf=false
7. ln_apr_obc=false
8. surgeonly - no tides, just surge and winds.
9. noPwSurge - no pressure forcing, but still has winds, surge and tides

In this notebook, I will compare the water levels from the model, observations and t_tide preditcions. I will also plot the model residuals and observed residuals. 

Model residuals = model run with all forcing - model run with tides only. 

Observed residuals = observed water level - t_tide prediction. I may think about a way to improve t_tide predictions. Right now I'm using t_xtide, but I could get better predictions by running t_pred on a year long time series...

### Error in using only 8 constituents

Since we are only using 8 tidal constituents, Rich suggested we examine that error. We can look at this by running t_predic on our time series with all constituents and then running it again using only 8 constituents.  I have a matlab script to do this: /data/nsoontie/MEOPAR/analysis/storm_surges/data/get_ttide_8.m.  This script saves a csv file with predictions using only 8 constituents and predictions using all constituents. Let's look at the results:

```python
run ='tidesonly'
key = 'PointAtkinson'

start = '31-Dec-2005'; end = '02-Jan-2007'
filename = '/data/nsoontie/MEOPAR/analysis/storm_surges/data/Point Atkinson_t_tide_compare8_' + start+'_'+end+'.csv'
def dateParserMeasured2(s):
    #convert the string to a datetime object
    unaware = datetime.datetime.strptime(s, "%d-%b-%Y %H:%M:%S ")
    #add in the local time zone (Canada/Pacific)
    aware = unaware.replace(tzinfo=pytz.timezone('Canada/Pacific'))
    #convert to UTC
    return aware.astimezone(pytz.timezone('UTC'))
    
fig = plt.figure(figsize=(20,10))
ax=pylab.subplot(1,1,1)
import matplotlib.dates as mdates
ax.fmt_xdata = mdates.DateFormatter('%Y-%m-%d')
fig.autofmt_xdate()

ttide = pd.read_csv(filename,delimiter='\t',skiprows=3,parse_dates=[0],date_parser=dateParserMeasured2)
ttide = ttide.rename(columns={'Time_Local ': 'time', ' pred_8 ': 'pred_8', ' pred_all ': 'pred_all'})
difference = ttide.pred_all-ttide.pred_8 # difference between the two
pylab.plot(ttide.time, ttide.pred_8, label='ttide with 8',lw=2)
pylab.plot(ttide.time, ttide.pred_all, label='ttide with all',lw=2)
pylab.plot(ttide.time, difference, label="Difference")
#pylab.plot(ts[run],sshs[run][key][:,0,0],'-',label='model')
ax.set_xlim([datetime.datetime(2006,2,2), datetime.datetime(2006,2,7)])
ax.set_ylim([-3,3])
pylab.legend(loc=0)
pylab.grid(True,lw=1)

print np.max(difference), np.min(difference)
```

So, we may underestimate peaks by using only 8 constituents. Over this time frame, the largest error is about 40cm on Feb 2.

Also, by using the 2005 time series in the harmonic analysis, the difference curve is shifted down by more than 5-10 cm. This is because the tidal predictions with many constituents is are smaller with the 2005 time series. The predictions with only 8 constituents are very similar with both the 2005 and 2006 time series.

** Conclusion**

We should use the 2005 harmonics to calculate tidal predictions.
