This notebook analyses a Feb 2006 storm surge with the corrected tides. K1 is phase/amp corrected, M2 is phase shifted, lowered bottom firction to 1e-3. Cases:

1. all_forcing - all tidal forcing and ssh forcing and meteorolgical forcing
2. tidesonly - no meteorological forcing and no ssh forcing. Only tidal forcing.

In this notebook, I will compare the water levels from the model, observations and t_tide preditcions. I will also plot the model residuals and observed residuals. 

Model residuals = model run with all forcing - model run with tides only. 

Observed residuals = observed water level - t_tide prediction. I may think about a way to improve t_tide predictions. Right now I'm using t_xtide, but I could get better predictions by running t_pred on a year long time series...



```python
%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np

import arrow
import datetime
import pandas as pd
import pytz
from salishsea_tools import tidetools
from salishsea_tools import stormtools
```

```python
path = '/data/nsoontie/MEOPAR/SalishSea/results/storm-surges/tide_fix/feb2006/'


runs = {'all_forcing','tidesonly', 'ssh_only'}

fUs={}; fVs={}; fTs={};

for key in runs:
    fUs[key] = NC.Dataset(path + key +'/SalishSea_4h_20060201_20060207_grid_U.nc','r');
    fVs[key] = NC.Dataset(path + key +'/SalishSea_4h_20060201_20060207_grid_V.nc','r');
    fTs[key] = NC.Dataset(path + key +'/SalishSea_4h_20060201_20060207_grid_T.nc','r');
```

```python
# define plot parameters
#time, depth, colormap etc
timestamp =41#model output time 
times = fVs[key].variables['time_counter'] #array of output times
time = times[timestamp]  #physical time in seconds
depthlevel = 0  #model level to plot
depths = fVs[key].variables['depthv'] #array of depths
depth = depths[depthlevel]  #physical depth in meters
cmap = plt.get_cmap('jet') 
u_lat = fUs[key].variables['nav_lat']
u_lon = fUs[key].variables['nav_lon']
T_lat = fTs[key].variables['nav_lat']
T_lon = fTs[key].variables['nav_lon']
v_lat = fVs[key].variables['nav_lat']
v_lon = fVs[key].variables['nav_lon']
```

```python
Us={}; Vs={}; Es ={}; Ss={}; Ts={};

for key in runs:
    [Us[key], Vs[key], Es[key], Ss[key], Ts[key]] = stormtools.get_variables(fUs[key],fVs[key],fTs[key],timestamp,depthlevel)
```

```python
start='01-Feb-2006'
stations = {'PointAtkinson': 7795, 'Victoria': 7120, 'PatriciaBay': 7277, 'CampbellRiver': 8074, 
            'CrescentBeach': 1, 'WhiteRock': 1, 'BoundaryBay': 1}
datums = {'PointAtkinson': 3.09, 'Victoria': 1.881, 'PatriciaBay': 2.256, 'CampbellRiver': 2.916, 
          'CrescentBeach': 2.44, 'WhiteRock': 2.85, 'BoundaryBay': 1 }
thalwegs = ['Thalweg1','Thalweg2', 'Thalweg3', 'Thalweg4', 'Thalweg5', 'Thalweg6']
allstations = {'PointAtkinson': 7795, 'Victoria': 7120, 'PatriciaBay': 7277, 'CampbellRiver': 8074,
               'CrescentBeach': 2.44, 'WhiteRock': 2.85 , 'BoundaryBay': 1,
               'Thalweg1': [], 'Thalweg2': [], 'Thalweg3': [], 'Thalweg4': [], 'Thalweg5': [], 'Thalweg6': [],
               'Plume': 1} 
#datums from t_xtide
run_stations={}
us={}; vs={}; lats={}; lons={}; tmps={}; sals={}; sshs={}; ts={};

for run in runs:
    for key in allstations:
        string = path + run + '/1h_' + key + '.nc'
        run_stations[key] = NC.Dataset(string,'r');
        t=run_stations[key].variables['time_counter']
        ts[run]=stormtools.convert_date_seconds(t,start)
    [us[run], vs[run], lats[run], lons[run], tmps[run], sals[run], sshs[run]] = stormtools.combine_data(run_stations)
    run_stations={};
```

```python


start='01-Feb-2006'
end='07-Feb-2006'
emin=-3
emax=3
plt.figure(figsize=(20,12))

wlev_pred_xtide={}
wlev_meas={}

i=1;
for key in stations:
    for run in runs:
        ax=pylab.subplot(4,2,i)
        pylab.plot(ts[run],sshs[run][key][:,0,0] ,label=run)
        pylab.title(key)
        ax.set_ylim([emin,emax])
        
     
    if key == 'PointAtkinson' or key == 'Victoria' or key =='PatriciaBay' or key =='CampbellRiver':
        #plotting observations
        statID_PA = stations[key]
        filename = 'wlev_' +str(statID_PA) + '_' + start +'_' +end +'.csv'
        tidetools.get_dfo_wlev(statID_PA,start,end)
        def dateParserMeasured(s):
            #convert the string to a datetime object
            unaware = datetime.datetime.strptime(s, "%Y/%m/%d %H:%M")
            #add in the local time zone (Canada/Pacific)
            aware = unaware.replace(tzinfo=pytz.timezone('Canada/Pacific'))
            #convert to UTC
            return aware.astimezone(pytz.timezone('UTC'))
        wlev_meas[key] = pd.read_csv(filename,skiprows=7,parse_dates=[0],date_parser=dateParserMeasured)
        wlev_meas[key] = wlev_meas[key].rename(columns={'Obs_date': 'time', 'SLEV(metres)': 'slev'})
        pylab.plot(wlev_meas[key].time, wlev_meas[key].slev-datums[key],'k*',label='observed') #subtracted off the datum from t_xtide

    
    if key != 'BoundaryBay':
        #plotting t_xtide predictions
        filename = '/data/nsoontie/MEOPAR/analysis/storm_surges/data/'+ key+ ' (2)_xtide_prediction_' + start+'_'+end+'.csv'
        def dateParserMeasured2(s):
            #convert the string to a datetime object
            unaware = datetime.datetime.strptime(s, "%d-%b-%Y %H:%M:%S ")
            #add in the local time zone (Canada/Pacific)
            aware = unaware.replace(tzinfo=pytz.timezone('Canada/Pacific'))
            #convert to UTC
            return aware.astimezone(pytz.timezone('UTC'))
        wlev_pred_xtide[key] = pd.read_csv(filename,delimiter='\t',parse_dates=[0],date_parser=dateParserMeasured2)
        wlev_pred_xtide[key] = wlev_pred_xtide[key].rename(columns={'Time_Local ': 'time', ' wlev_pred ': 'slev'})
        #pylab.plot(wlev_pred_xtide[key].time, wlev_pred_xtide[key].slev-datums[key], label='t_xtide')
    
    pylab.grid()
    i=i+1

pylab.legend(bbox_to_anchor=(1.05, 1), loc=2)
print 'Sea Surface Height'
```

### Error in using only 8 constituents

Since we are only using 8 tidal constituents, Rich suggested we examine that error. We can look at this by running t_predic on our time series with all constituents and then running it again using only 8 constituents.  I have a matlab script to do this: /data/nsoontie/MEOPAR/analysis/storm_surges/data/get_ttide_8.m.  This script saves a csv file with predictions using only 8 constituents and predictions using all constituents. Let's look at the results:

```python
run ='tidesonly'
key = 'PointAtkinson'

start = '31-Dec-2005'; end = '02-Jan-2007'
filename = '/data/nsoontie/MEOPAR/analysis/storm_surges/data/Point Atkinson_t_tide_compare8_' + start+'_'+end+'.csv'
def dateParserMeasured2(s):
    #convert the string to a datetime object
    unaware = datetime.datetime.strptime(s, "%d-%b-%Y %H:%M:%S ")
    #add in the local time zone (Canada/Pacific)
    aware = unaware.replace(tzinfo=pytz.timezone('Canada/Pacific'))
    #convert to UTC
    return aware.astimezone(pytz.timezone('UTC'))
    
fig = plt.figure(figsize=(20,10))
ax=pylab.subplot(1,1,1)
import matplotlib.dates as mdates
ax.fmt_xdata = mdates.DateFormatter('%Y-%m-%d')
fig.autofmt_xdate()

ttide = pd.read_csv(filename,delimiter='\t',parse_dates=[0],date_parser=dateParserMeasured2)
ttide = ttide.rename(columns={'Time_Local ': 'time', ' pred_8 ': 'pred_8', ' pred_all ': 'pred_all'})
pylab.plot(ttide.time, ttide.pred_8, label='t_predic with 8',lw=2)
pylab.plot(ttide.time, ttide.pred_all, 'k', label='t_predic with all',lw=2)
#pylab.plot(ts[run],sshs[run][key][:,0,0],'-',label='model')
ax.set_xlim([datetime.datetime(2006,2,2), datetime.datetime(2006,2,7)])
pylab.legend(loc=0)
pylab.grid(True,lw=1)
```

So, we may underestimate peaks by using only 8 constituents. Over this time frame, the largest error is about 40cm on Feb 2.

```python
def plot_forcing(key, start, end, start_int, end_int):

    depth=0; 

    #some empty arrays for later
    pres={}; windspeed={};time={}

    #plotting details
    wind_ax=[0,24]; pres_ax=[94000,102000]; ax_ssh=[-3,3]; ax_cur=[-2,2]; ax_sal=[10,36]

    #retreive CGRF and observed weather
    [windspeed['CGRF'],pres['CGRF'],time['CGRF']]=stormtools.get_CGRF_weather(start,end,CGRF_points[key])
    [windspeed['obs'],time['obs']] =stormtools.get_EC_observations(key, start,end)

    #plotting
    plt.figure(figsize=(8,18))
    #winds
    ax=pylab.subplot(6,1,1)
    pylab.plot(time['CGRF'],windspeed['CGRF'],'g',label='CGRF')
    pylab.plot(time['obs'],windspeed['obs'],'k',label='Observations')
    pylab.title('Wind Speed')
    pylab.legend(bbox_to_anchor=(1.05, 1), loc=2)
    ax.set_ylim(wind_ax)
    pylab.grid()

    #pressure
    ax=pylab.subplot(6,1,2)
    pylab.plot(time['CGRF'],pres['CGRF'],'g',label='CGRF')
    pylab.title('Pressure')
    pylab.legend(bbox_to_anchor=(1.05, 1), loc=2)
    pylab.grid()
    ax.set_ylim(pres_ax)

    #ssh forcing
    ax=pylab.subplot(6,1,3)
    pylab.plot(time_ssh['Tofino'],ssh_forc['Tofino'][:,0,0],'g',label='SSH Anomaly at Tofino')
    pylab.plot(time_ssh['Port Hardy'],ssh_forc['Port Hardy'][:,0,0],'y',label='SSH Anomaly at Port Hardy')
    pylab.title('SSH Anomoaly Forcing')
    ax.set_xlim(['Dec 11, 2006', 'Dec 18, 2006'])
    pylab.legend(bbox_to_anchor=(1.05, 1), loc=2)
    pylab.grid()
    ax.set_ylim([-1,1])

    #plot model ssh with observations and tidal predictions
    ax=pylab.subplot(6,1,4)
    for run in runs:
        pylab.plot(ts[run],sshs[run][key][:,0,0],label=run)
    #pylab.plot(wlev_meas[key].time,wlev_meas[key].slev,'k',label='Observations')
    #pylab.plot(wlev_pred_xtide[key].time, wlev_pred_xtide[key].slev,'m', label='t_xtide')
    pylab.title('SSH')
    pylab.legend(bbox_to_anchor=(1.05, 1), loc=2)
    pylab.grid()
    ax.set_ylim(ax_ssh)

    #plot model currents
    ax=pylab.subplot(6,1,5)
    for run in runs:
        pylab.plot(ts[run],np.sqrt(us[run][key][:,depth,0,0]**2 + vs[run][key][:,depth,0,0]**2),label=run)
    pylab.title('Surface currents')
    pylab.grid(); ax.set_ylim(ax_cur)
    pylab.legend(bbox_to_anchor=(1.05, 1), loc=2)

    #plot model surface salinity
    ax=pylab.subplot(6,1,6)
    for run in runs:
        pylab.plot(ts[run],sals[run][key][:,depth,0,0],label=run)
    pylab.title('Surface salinity')
    pylab.grid(); ax.set_ylim(ax_sal)
    pylab.legend(bbox_to_anchor=(1.05, 1), loc=2)
```
