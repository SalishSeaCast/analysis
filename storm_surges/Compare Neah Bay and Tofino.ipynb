This notebook compares the Feb 2006 surge at Neah Bay and Tofino to determine if the timing of the surge event at these two location is in sync.

Also does a quick check of the contribution to SSH due to the inverse barometer effect.

# Neah Bay

Tidal predictions and water level observations are taken from the NOAA website. 
http://tidesandcurrents.noaa.gov/waterlevels.html?id=9443090&units=metric&bdate=20060201&edate=20060207&timezone=GMT&datum=MLLW&interval=h&action=data

The surge is calculated as the difference between observations and predictions.

Note: It was not completely obvious how to collect historical surge data from the National Weather Service website.
http://www.nws.noaa.gov/mdl/etsurge/index.php?page=stn&region=wc&datum=msl&list=&map=0-48&type=both&stn=waneah



# Tofino

Water level observations are taken from DFO. Tidal predictions are calculated using t_tide with tidal harmonics from a 2006 time series. The 2006 mean sea level from observations is added to the tidal predictions. Note: It is likely better to calculate harmonics using the 2005 time series. The most up to date files use the 2006 time series.   

The surge is calculated as the difference between observations and predictions.

```python

%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np

import arrow
import datetime
import pandas as pd
import pytz
from salishsea_tools import tidetools
from salishsea_tools import stormtools
```

```python
start='01-Feb-2006'
end = '07-Feb-2006'
```

First, load Neah Bay data.

```python


filename_wl = 'data/NeahBay_' +start +'_' +end +'_wl.csv';
filename_pr = 'data/NeahBay_' +start +'_' +end +'_pr.csv';

def dateParserMeasured(s):
    #convert the string to a datetime object
    unaware = datetime.datetime.strptime(s, "%Y-%m-%d %H:%M")
    #add in the local time zone (Canada/Pacific)
    aware = unaware.replace(tzinfo=pytz.timezone('UTC'))
    #conve to UTC
    return aware.astimezone(pytz.timezone('UTC'))
    
NB_meas = pd.read_csv(filename_wl,skiprows=0,parse_dates=[0],date_parser=dateParserMeasured)
NB_meas = NB_meas.rename(columns={'Date Time': 'time', ' Water Level': 'slev', 
                                              ' Sigma': 'sigma', 'I': 'I', 'L': 'L'})

NB_pred = pd.read_csv(filename_pr,skiprows=0,parse_dates=[0],date_parser=dateParserMeasured)
NB_pred = NB_pred.rename(columns={'Date Time': 'time', ' Prediction': 'pred'})
```

Caclulate anomaly at Neah Bay

```python
NB_anom = np.zeros(len(NB_meas.time))

for i in np.arange(0,len(NB_meas.time)):
    #check that there is a corresponding time 
    #if any(wlev_pred.time == wlev_meas.time[i]):
    NB_anom[i] =(NB_meas.slev[i] - NB_pred.pred[NB_pred.time==NB_meas.time[i]])
    if not(NB_anom[i]):
        NB_anom[i]=float('Nan')
```

Load in Tofino surge.

```python
ssh_Tofino = '/data/nsoontie/MEOPAR/NEMO-forcing/open_boundaries/west/ssh/ssh'

date='01-Feb-2006'


arr_start =arrow.Arrow.strptime(date,'%d-%b-%Y')
yr=arr_start.year
m=arr_start.month
if m<10:
    m='0' + str(m)
string = '_y'+ str(yr) + 'm' + str(m) +'.nc'
    
 #get the data
fT = NC.Dataset(ssh_Tofino + string);
Tof_anom=fT.variables['sossheig'];
tss= fT.variables['time_counter'][:];
l = tss.shape[0]; t=np.linspace(0,l-1,l) #time array
Tof_time=stormtools.convert_date_hours(t,date);
```

```python
print Tof_anom.shape
```

```python
#plot everything now
plt.figure(figsize=(18,5))
ax1=pylab.subplot(1,1,1)

pylab.plot(NB_meas.time,NB_anom,'b',label='Neah Bay')
pylab.plot(Tof_time[:],Tof_anom[:,0,0],'r',label='Tofino')

ax1.set_ylim([-1,1])
ax1.set_xlim([datetime.datetime(2006,2,1), datetime.datetime(2006,2,7)])

pylab.xlabel('Time')
pylab.ylabel('SSH anomaly (m)')
pylab.grid()
pylab.legend(loc=0)
pylab.title('SSH Anomaly for ' +date)
    
```

The timing and amplitude of the surge between Neah Bay and Tofino is similar. Neah Bay peak is slightly later (2 hours). 

Also, Neah Bay has a slightly higher anomaly during 'non-surge' times. 

```python
#find timing of max at both Tofino and Neah Bay

max_NB = np.max(NB_anom)
time_NB = NB_meas.time[NB_anom.argmax()]

max_Tof= np.max(Tof_anom[:,0,0])
time_Tof = Tof_time[Tof_anom[:,0,0].argmax()]

print 'Max at Neah Bay:', max_NB, 'Time:', time_NB
print 'Max at Tofino:', max_Tof, 'Time:', time_Tof
```

# Inverse Barometer

NEMO can correct the sea surface height to account for the force applied the atmospheric pressure. If the atmosphere is at a low pressure then the ssh will be elevated slightly. 

The correction fator given by the NEMO documentation is:

$ \eta_{ib} = -\frac{1}{g\rho_0}\left(P_{atm}-P_0\right) $
