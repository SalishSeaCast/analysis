This notebook looks for a grid point close to Delta so that we can start including that community in our output. 

Unfortunately there are no permanent tide gauges in this region so we won't be able to compare with observations. DFO produces tide charts for Tsawwassen.

There are xtide stations for Crescent Beach and White Rock. I will add these two locations to out storm outputs. 

```python
%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np
import arrow
from salishsea_tools import tidetools
```

First, I have identified some regions near Delta that were affected by storm surge flooding. Boundary Bay is a big one. Using Google Maps I have located the lat/long of Boundary Bay.

Crescent Beach and White Rock coordinates taken from xtide website.
http://tbone.biol.sc.edu/tide/tideshow.cgi

NOAA site for American stations
http://tidesandcurrents.noaa.gov/map/index.shtml?type=PreliminaryData&region=Washington

```python
BB_lat = 49.021131 #Bounday Bay latitude
BB_lon = -123.041758 #Bounday Bay longitude

WR_lat= 49.0167 #White Rock Latitude
WR_lon = -122.8 #White Rock Longitude

CB_lat=49.0333 #Crescent Beach Longitude	
CB_lon=-122.8833 #Crescent Beach Latitude

NB_lat=48.37 #Neah Bay latitude
NB_lon=-124.6 #Neah Bay longitude (Taken from NOAA)

CP_lat=48.87 #Cherry Point lat
CP_lon=-122.75 #Cheery Point long (taken from NOAA)

FH_lat = 48.55 #Friday Harbor lat
FH_lon = -123.02 #Friday Harbor lon (taken from NOAA)

PA_lat = 48.13 #Port Angeles lat
PA_lon = -123.42 #Port Angeles lon (taken from NOAA)
```

Now load the grid and bathymetry.

```python
grid = NC.Dataset('../../NEMO-forcing/grid/bathy_meter_SalishSea2.nc')
bathy = grid.variables['Bathymetry'][:]
X=grid.variables['nav_lon'][:]
Y=grid.variables['nav_lat'][:]
```

```python
[i,j] = tidetools.find_closest_model_point(BB_lon,BB_lat,X,Y,bathy)

[iWR,jWR] = tidetools.find_closest_model_point(WR_lon,WR_lat,X,Y,bathy)

[iCB,jCB] = tidetools.find_closest_model_point(CB_lon,CB_lat,X,Y,bathy)

[iNB,jNB] = tidetools.find_closest_model_point(NB_lon,NB_lat,X,Y,bathy)

[iCP,jCP] = tidetools.find_closest_model_point(CP_lon,CP_lat,X,Y,bathy)

[iFH,jFH] = tidetools.find_closest_model_point(FH_lon,FH_lat,X,Y,bathy)

[iPA,jPA] = tidetools.find_closest_model_point(PA_lon,PA_lat,X,Y,bathy)
```

Lets plot to make sure that the coodinates are where we would expect. For instance, i think the i index correponds to the vertical axis and the j index corresponds the the horizontal axis.

```python
plt.figure(figsize=(16,8))
pylab.subplot(1,3,1)
pylab.pcolormesh(bathy)
pylab.plot(j,i,'or',label='Boundary Bay')
pylab.plot(jWR,iWR,'og',label='White Rock')
pylab.plot(jCB,iCB,'oc',label='Crescent Beach')
pylab.plot(jNB,iNB,'om',label='Neah Bay')
pylab.plot(jCP,iCP,'ob',label='Cherry Point')
pylab.plot(jFH,iFH,'oy',label='Friday Harbor')
pylab.plot(jPA,iPA,'ok',label='Port Angeles')
pylab.legend(loc=0)

pylab.subplot(1,3,2)
pylab.pcolormesh(bathy)
pylab.plot(j,i,'or', label='Boundary Bay')
pylab.plot(jWR,iWR,'og',label='White Rock')
pylab.plot(jCB,iCB,'oc',label='Crescent Beach')
pylab.plot(jCP,iCP,'ob',label='Cherry Point')
pylab.plot(jFH,iFH,'oy',label='Friday Harbor')
pylab.legend(loc=0)
pylab.axis([250,360,250,400])

pylab.subplot(1,3,3)
pylab.pcolormesh(bathy)
pylab.plot(jNB,iNB,'om',label='Neah Bay')
pylab.plot(jPA,iPA,'ok',label='Port Angeles')
pylab.legend(loc=0)
pylab.axis([0,180,240,470])
```

This looks like the right spot, but I will add on to the horizontal index so that it is touching right against the land. Then hopefully the other grids won't be masked there. Also add one to account for NEMO fortran indexing.

Neah Bay is right in a corner too. Again, add one teach index to get away from corner.

```python
#Adjustments for grids
#Boundary Bay: Move right one
print 'Printing fortran indices'
xind= j+1;
yind=i;
print 'Boundary Bay'
print xind+1,yind+1

#White Rock: Move left one
xWR = jWR-1
yWR=iWR
print 'White Rock'
print xWR+1,yWR+1

#Crescent Beach: Move left one
xCB = jCB-1
yCB=iCB
print 'Crescent Beach'
print xCB+1,yCB+1

#Neah Bay: Move right one, move up one
xNB = jNB+1
yNB=iNB+1
print 'Neah Bay'
print xNB+1,yNB+1

#Cherry Point: leave
xCP = jCP
yCP=iCP
print 'Cherry Point'
print xCP+1,yCP+1

#Friday Harbor: leave
xFH = jFH
yFH=iFH
print 'Friday Harbor'
print xFH+1,yFH+1

#Port Angeles: leave
xPA = jPA
yPA=iPA+1
print 'Port Angeles'
print xPA+1,yPA+1
```
