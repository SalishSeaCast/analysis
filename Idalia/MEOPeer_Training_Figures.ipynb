```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.mlab as mlab
from matplotlib import pyplot as plt
from matplotlib import animation
from salishsea_tools import stormtools

import netCDF4 as nc
from salishsea_tools import (
    nc_tools,
    viz_tools,)
import matplotlib.gridspec as gridspec

%matplotlib inline
```

```python
tracers = nc.Dataset('/ocean/imachuca/MEOPAR/data/results/plume/jun2003/SalishSea_1h_20030531_20030602_grid_T.nc')
tracersU = nc.Dataset('/ocean/imachuca/MEOPAR/data/results/plume/jun2003/SalishSea_1h_20030531_20030602_grid_U.nc')
tracersV = nc.Dataset('/ocean/imachuca/MEOPAR/data/results/plume/jun2003/SalishSea_1h_20030531_20030602_grid_V.nc')
timesteps = tracers.variables['time_counter']

bathy = nc.Dataset('/ocean/imachuca/MEOPAR/NEMO-forcing/grid/bathy_meter_SalishSea2.nc')
ugrid = tracersU.variables['vozocrtx']; ugridshape = ugrid.shape
vgrid = tracersV.variables['vomecrty']
zlevels = tracersU.variables['depthu']
timesteps = tracersU.variables['time_counter']
salvar = tracers.variables['vosaline']

sal0 = salvar[:,0,:,:]

cmin = sal0.min()
cmax = sal0.max()
```

```python
#Setting up a blank figure
fig = plt.figure(figsize=(20,15))
gs = gridspec.GridSpec(2, 2, width_ratios=[1,1], height_ratios=[2,1])
#gs.update(wspace=0.15, hspace=0.1)
ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,:])


viz_tools.set_aspect(ax1)
viz_tools.set_aspect(ax2)
cmap = plt.get_cmap('Blues')
viz_tools.plot_land_mask(ax1, bathy, color='burlywood')
viz_tools.plot_land_mask(ax2, bathy, color='burlywood')

def init():
    sal = np.ma.masked_values(sal0[0], 0)
    mesh = ax1.pcolormesh(sal,cmap=cmap)
    mesh = ax2.pcolormesh(sal,cmap=cmap)
    mesh.set_clim(cmin,cmax)
    cbar = fig.colorbar(mesh,ax=ax1)
    cbar = fig.colorbar(mesh,ax=ax2)
    cbar.set_label('{label} [{units}]'.format(label=salvar.long_name.title(), units=salvar.units))
    ax1.set_xlabel('x Index')
    ax1.set_ylabel('y Index')
    ax1.set_title('t = 0.0 h')
    ax2.set_xlabel('x Index')
    ax2.set_ylabel('y Index')
    ax2.set_title('t = 0.0 h')


def animate(t):
    ssh = np.ma.masked_values(sal0[t], 0)
    mesh = ax1.pcolormesh(ssh,cmap=cmap)
    mesh = ax2.pcolormesh(ssh,cmap=cmap)
    mesh.set_clim(cmin,cmax)
    ax1.set_xlabel('x Index')
    ax1.set_ylabel('y Index')
    ax2.set_xlabel('x Index')
    ax2.set_ylabel('y Index')
    timestamp = nc_tools.timestamp(tracers,t)
    ax1.set_title(timestamp)
    ax2.set_title(timestamp)
    
    xstart=0
    xend=ugridshape[3]
    ystart=200
    yend=600
    y_slice = np.arange(ystart, yend)
    x_slice = np.arange(xstart, xend)
    arrow_step = 25
    y_slice_a = y_slice[::arrow_step]
    x_slice_a = x_slice[::arrow_step]
    ugrid_tzyx = np.ma.masked_values(ugrid[t, zlevel, y_slice_a, x_slice_a], 0)
    vgrid_tzyx = np.ma.masked_values(vgrid[t, zlevel, y_slice_a, x_slice_a], 0)
    u_tzyx, v_tzyx = viz_tools.unstagger(ugrid_tzyx, vgrid_tzyx)
    speeds = np.sqrt(np.square(u_tzyx) + np.square(v_tzyx))
    

#The animation function    
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=2, interval=300, blit=False, repeat=False)

#A line that makes it all work
mywriter = animation.FFMpegWriter()

#Save in current folder
anim.save('Training_Tracers_Animation.mp4',writer=mywriter,fps=1)

```
