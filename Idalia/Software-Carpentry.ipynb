```python
lon=data.variables['traj_lon']
lat=data.variables['traj_lat']
dep=data.variables['traj_depth']
x=data.variables['init_x']
y=data.variables['init_y']
```

# Plotting (2D)

```python
fig, ax = plt.subplots(1,1,figsize=(20,10))
ax.scatter(lon[:,0],lat[:,0],color='blue',label='Particle 1')
ax.legend()

ax.set_title('2D')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
```

# Plotting (3D)

```python
fig = plt.figure(figsize=(20,10))
ax = fig.gca(projection='3d')
ax.scatter(lon[:,0],lat[:,0],dep[:,0],color='blue',label='Particle 1')

ax.set_title('3D')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_zlabel('Depth')
```

# Plotting (3D): Camera Position

```python
fig = plt.figure(figsize=(20,10))
ax = fig.gca(projection='3d')
ax.scatter(lon[:,0],lat[:,0],dep[:,0],color='blue',label='Particle 1')

ax.set_title('3D')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_zlabel('Depth')

ax.view_init(elev=0, azim=-90)
```
