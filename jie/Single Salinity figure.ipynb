```python
saline=tracers.variables['vosaline']
zlevels=tracers.variables['deptht']
timesteps=tracers.variables['time_counter']
lons=tracers.variables['nav_lon']
lats=tracers.variables['nav_lat']
```

```python
t, zlevel = 3, 0
saline_tz = np.ma.masked_values(saline[t, zlevel], 0)

fig, ax = plt.subplots(1, 1, figsize=(10, 8))
viz_tools.set_aspect(ax)
cmap = plt.get_cmap('spectral')
cmap.set_bad('burlywood')
mesh = ax.pcolormesh(saline_tz, cmap=cmap)
cbar = fig.colorbar(mesh)
plt.axis((0, saline_tz.shape[1], 0, saline_tz.shape[0]))

ax.grid()
ax.set_xlabel('x Index')
ax.set_ylabel('y Index')
cbar.set_label('Practical Salinity')
ax.set_title(u't = {t:.1f}h, depth \u2248 {d:.2f}{z.units}'.format(t=timesteps[t] / 3600, d=zlevels[zlevel], z=zlevels))
```

```python
fig,ax=plt.subplots(1,1,figsize=(10,8))
viz_tools.set_aspect(ax,coords='map',lats=lats)
cmap=plt.get_cmap('spectral')
cmap.set_bad('burlywood')
mesh=ax.pcolormesh(lons[:],lats[:],saline_tz[:],cmap=cmap)
cbar=fig.colorbar(mesh)

ax.set_xlabel('{longitude.long_name} [{longitude.units}]'.format(longitude=lons))
ax.set_ylabel('{latitude.long_name} [{latitude.units}]'.format(latitude=lats))
cbar.set_label('Practical Salinity')
```

```python
t, zlevel = 3, 0
y_map=np.arange(400,450)
x_map=np.arange(275,325)
lons_map=lons[y_map,x_map]
lats_map=lats[y_map,x_map]
saline_tz = np.ma.masked_values(saline[t, zlevel,400:450,275:325], 0)

fig,ax=plt.subplots(1,1,figsize=(10,8))
viz_tools.set_aspect(ax,coords='map',lats=lats)
cmap=plt.get_cmap('spectral')
cmap.set_bad('burlywood')
mesh=ax.pcolormesh(lons_map,lats_map,saline_tz[:],cmap=cmap)
clines = ax.contour(lons_map, lats_map, saline_tz, np.arange(2, 20))
ax.clabel(clines,fmt='%2.1f')
cbar = fig.colorbar(mesh, shrink=0.60)
plt.plot(lon1,regr.predict(lon1),'b',linewidth=1)

ax.set_xlabel('{longitude.long_name} [{longitude.units}]'.format(longitude=lons))
ax.set_ylabel('{latitude.long_name} [{latitude.units}]'.format(latitude=lats))
cbar.set_label('Practical Salinity')
```
