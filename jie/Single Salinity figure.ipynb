```python
from __future__ import division, print_function

import matplotlib.pyplot as plt
import netCDF4 as nc
import numpy as np

from salishsea_tools import (
    nc_tools,
    viz_tools,
)
```

```python
%matplotlib inline
```

```python
tracers = nc.Dataset('/ocean/jieliu/research/meopar/SalishSea_1h_20140910_20140910_grid_T.nc')
```

```python
nc_tools.show_dimensions(tracers)
```

```python
nc_tools.show_variables(tracers)
```

```python
nc_tools.show_variable_attrs(tracers, 'time_counter')
tracers.variables['time_counter'][:]
```

```python
nc_tools.show_variable_attrs(tracers, 'time_counter_bnds')
tracers.variables['time_counter_bnds'][:]
```

```python
timesteps = tracers.variables['time_counter']
```

```python
nc_tools.show_variable_attrs(tracers, 'vosaline')
```

```python
saline = tracers.variables['vosaline']
```

```python
print(saline[0].shape)
#saline[0]
```

```python
saline_part = saline[0,0, 0:898, 0:398]
print(saline_part.shape)
saline_part
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
viz_tools.set_aspect(ax)
mesh = ax.pcolormesh(saline_part)
fig.colorbar(mesh)
```

```python
saline0 = np.ma.masked_values(saline_part, 0)
print(saline0.shape)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
viz_tools.set_aspect(ax)
mesh = ax.pcolormesh(saline0)
fig.colorbar(mesh)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
viz_tools.set_aspect(ax)
cmap = plt.get_cmap('jet')
cmap.set_bad('burlywood')
mesh = ax.pcolormesh(saline0, cmap=cmap)
cbar = fig.colorbar(mesh)
plt.axis((0, saline0.shape[1], 0, saline0.shape[0]))

ax.set_xlabel('x Index')
ax.set_ylabel('y Index')
cbar.set_label('{label} [{units}]'.format(label=saline.long_name.title(), units=saline.units))
```

```python
saline1=saline[0] #different construction from which is in SSH
print(saline1.shape)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(16, 8), sharey=True)
cmap = plt.get_cmap('jet')
cmap.set_bad('burlywood')
time_steps = (0,4,8)
for ax, t in zip(axs, time_steps):
    saline_t = np.ma.masked_equal(saline1[t], 0)
    viz_tools.set_aspect(ax)
    mesh = ax.pcolormesh(saline_t, cmap=cmap)
    cbar = fig.colorbar(mesh, ax=ax)
    ax.set_title('t = {:.1f}h'.format(timesteps[t] / 3600))
    ax.set_xlabel('x Index')
    
axs[0].set_ylabel('y Index')
cbar.set_label('{label} [{units}]'.format(label=saline.long_name.title(), units=saline.units))
    
```
